```python
# @title 🌳 [CELL 3B] The GEDCOM & Naming Engine
# 🛑 DOCTRINE LOCK: YATES DNA ANCHOR STUDY
# RULE: ID-TO-GEDCOM NAMING ONLY.
# RULE: CSV IS TRUTH (CSS/DOCS); GEDCOM IS MAP.
# RULE: GENERATOR-FIRST (NO SURGICAL SERVER PATCHES).
# RULE: NO PARTIAL BLOCKS. COMPILE FULL CELL.

print("="*60)
print("      [PHASE 2] NATIVE GEDCOM PARSING & NAMING MATH...")
print("="*60)

# --- STEP 3: LOCALIZE & PREP GEDCOM ---
ged_files = [f for f in glob.glob("*.ged") if "_processed" not in f.lower()]
if not ged_files: raise FileNotFoundError("❌ No native GEDCOM found. Please ensure a .ged file is in the directory.")
INPUT_GED = sorted(ged_files, key=lambda x: os.path.getmtime(x), reverse=True)[0]
print(f"    [+] Active Native GEDCOM Map: {INPUT_GED}")

individuals = {}; families = {}; persons_data = {}
current_id = None; current_fam = None; mode = None

with open(INPUT_GED, "r", encoding="utf-8", errors="replace") as f:
    for line in f:
        line = line.strip(); parts = line.split(" ", 2)
        if len(parts) < 2: continue
        lvl, tag, val = parts[0], parts[1], parts[2] if len(parts)>2 else ""

        if lvl == "0" and "INDI" in val:
            current_id = tag.replace("@", "")
            if not current_id.startswith('I'): current_id = 'I' + current_id
            individuals[current_id] = {"name": "findme", "famc": None, "fams": [], "sex": "U", "raw_name": ""}
            persons_data[current_id] = {'name': f'ID {current_id}', 'bdate': '', 'bplace': '', 'ddate': '', 'dplace': '', 'mdate': '', 'sources_count': 0, 'citations_count': 0, 'raw_note': ''}
            mode = None
        elif current_id and lvl != "0":
            if tag == "NAME" and lvl == "1":
                individuals[current_id]["raw_name"] = val
                n = val.replace("/", "").strip()
                individuals[current_id]["name"] = n if n.lower() not in ["unknown","living"] else "findme"
                persons_data[current_id]["name"] = n
            elif tag == "SEX" and lvl == "1":
                individuals[current_id]["sex"] = val.strip()
            elif tag == "FAMC" and lvl == "1":
                individuals[current_id]["famc"] = val.replace("@", "")
            elif tag == "FAMS" and lvl == "1":
                individuals[current_id]["fams"].append(val.replace("@", ""))
            elif tag == "BIRT": mode = "BIRT"
            elif tag == "DEAT": mode = "DEAT"
            elif tag == "NOTE" and lvl == "1":
                persons_data[current_id]['raw_note'] = persons_data[current_id].get('raw_note', '') + " " + val
                mode = "NOTE"
            elif tag == "CONT" and mode == "NOTE": persons_data[current_id]['raw_note'] += " " + val
            elif tag == "CONC" and mode == "NOTE": persons_data[current_id]['raw_note'] += val
            elif tag == "SOUR" and lvl == "1": persons_data[current_id]['sources_count'] += 1; mode = None
            elif tag == "SOUR" and lvl == "2": persons_data[current_id]['citations_count'] += 1
            elif tag == "DATE" and mode == "BIRT": persons_data[current_id]['bdate'] = val
            elif tag == "DATE" and mode == "DEAT": persons_data[current_id]['ddate'] = val
            elif tag == "PLAC" and mode == "BIRT": persons_data[current_id]['bplace'] = val
            elif tag == "PLAC" and mode == "DEAT": persons_data[current_id]['dplace'] = val
            elif lvl == "1": mode = None
        elif lvl == "0" and "FAM" in val:
            current_fam = tag.replace("@", "")
            families[current_fam] = {"husb": None, "wife": None, "mdate": ""}
            current_id = None
            mode = None
        elif current_fam and lvl != "0":
            if tag == "HUSB":
                husb_id = val.replace("@", "")
                if not husb_id.startswith('I'): husb_id = 'I' + husb_id
                families[current_fam]["husb"] = husb_id
            elif tag == "WIFE":
                wife_id = val.replace("@", "")
                if not wife_id.startswith('I'): wife_id = 'I' + wife_id
                families[current_fam]["wife"] = wife_id
            elif tag == "MARR": mode = "MARR"
            elif tag == "DATE" and mode == "MARR": families[current_fam]["mdate"] = val
            elif lvl == "1": mode = None

for fid, f in families.items():
    if f.get("mdate"):
        if f["husb"] and f["husb"] in persons_data: persons_data[f["husb"]]["mdate"] = f["mdate"]
        if f["wife"] and f["wife"] in persons_data: persons_data[f["wife"]]["mdate"] = f["mdate"]

# --- STEP 3.5: COMPUTATIONAL NAMING ENGINE (ASCII PIVOT APPLIED) ---
print("    [+] Formatting Names (Applying Spanish/ASCII nee Math to ALL Nodes)...")

def get_sur(pid):
    if pid in override_dict and override_dict[pid]['surname']: return override_dict[pid]['surname']
    if not pid or pid not in individuals: return ""
    rn = individuals[pid].get("raw_name", "")
    if "/" in rn: return rn.split("/")[1].strip()
    return rn.split()[-1] if len(rn.split()) > 1 else ""

def get_giv(pid):
    if pid in override_dict and override_dict[pid]['given']: return override_dict[pid]['given']
    if not pid or pid not in individuals: return ""
    rn = individuals[pid].get("raw_name", "")
    if "/" in rn: giv = rn.split("/")[0].strip()
    else: giv = " ".join(rn.split()[:-1]) if len(rn.split()) > 1 else rn
    return re.sub(r'\d+', '', giv).strip()

for pid, p in individuals.items():
    if p["name"] == "findme": continue
    giv = get_giv(pid); sur = get_sur(pid)
    d_id = None; m_id = None
    famc = p.get("famc")
    if famc and famc in families:
        d_id = families[famc].get("husb"); m_id = families[famc].get("wife")

    d_sur = get_sur(d_id).split('-')[0].strip() if d_id else ""
    m_sur = get_sur(m_id).split('-')[0].strip() if m_id else ""
    maiden_sur = f"{d_sur}-{m_sur}" if d_sur and m_sur and d_sur!="findme" and m_sur!="findme" else d_sur if d_sur and d_sur!="findme" else m_sur if m_sur and m_sur!="findme" else ""

    sex = p.get("sex", "U"); h_sur = ""
    if sex == "F" and p.get("fams"):
        for fid in p["fams"]:
            if fid in families:
                h_id = families[fid].get("husb")
                h_s = get_sur(h_id).split('-')[0].strip() if h_id else ""
                if h_s and h_s != "findme": h_sur = h_s; break

    final_married_sur = h_sur if h_sur else sur
    # 🚨 ASCII PIVOT: 'née' replaced with 'nee' for system stability
    if sex == "F" and maiden_sur and final_married_sur.lower() != maiden_sur.lower() and not giv.endswith(maiden_sur):
        display_name = f"{giv} {final_married_sur}, nee {maiden_sur}".strip()
        p["gedcom_name"] = f"{giv} /{final_married_sur}/, nee {maiden_sur}".strip()
    else:
        use_sur = f"{sur}-{m_sur}" if maiden_sur and m_sur and sur.lower() != d_sur.lower() and d_sur else maiden_sur if maiden_sur else sur
        display_name = f"{giv} {use_sur}".strip()
        p["gedcom_name"] = f"{giv} /{use_sur}/".strip()

    p["name"] = display_name
    if pid in persons_data: persons_data[pid]["name"] = display_name
    if pid in csv_auth: csv_auth[pid]["name"] = display_name

print("    ✅ Phase 2 Complete. GEDCOM structured and ASCII 'nee' naming universally applied.")
```

```python
# @title ⚙️ [CELL 3C] The ANCHOR Math & Network Demographics (Topology Upgrade)
# 🛑 DOCTRINE LOCK: YATES DNA ANCHOR STUDY
# RULE: ID-TO-GEDCOM NAMING ONLY.
# RULE: CSV IS TRUTH (CSS/DOCS); GEDCOM IS MAP.
# RULE: GENERATOR-FIRST (NO SURGICAL SERVER PATCHES).
# RULE: NO PARTIAL BLOCKS. COMPILE FULL CELL.

print("="*60)
print("      [PHASE 3] EXECUTING ANCHOR MATH & DEMOGRAPHICS...")
print("="*60)

# --- STEP 4: LINEAGE EXTRACTION ---
print("    [+] Tracing Lineages on the Map...")
def get_parents(pid):
    if not pid or pid not in individuals: return None, None
    famc = individuals[pid]["famc"]
    if not famc or famc not in families: return None, None
    return families[famc]["husb"], families[famc]["wife"]

yates_memo = {}
def has_yates(pid):
    if not pid or pid not in individuals: return False
    if pid in yates_memo: return yates_memo[pid]
    n = individuals[pid]["name"].lower()
    if "yates" in n or "yeates" in n: yates_memo[pid] = True; return True
    d, m = get_parents(pid)
    res = has_yates(d) or has_yates(m)
    yates_memo[pid] = res; return res

def climb(start_id):
    curr = start_id; lin = []
    while curr:
        p = individuals.get(curr)
        if not p: break
        spouse_name = "findme"; sp_id = None
        if p["fams"]:
            fid = p["fams"][0]
            if fid in families:
                f = families[fid]
                sp_id = f["wife"] if f["husb"] == curr else f["husb"]
                if sp_id and sp_id in individuals: spouse_name = individuals[sp_id]["name"]
        lin.append({"name": p["name"], "id": curr, "spouse": spouse_name, "sp_id": sp_id})
        d, m = get_parents(curr)
        if not d and not m: break
        dy, my = has_yates(d), has_yates(m)
        curr = d if dy and not my else (m if my and not dy else (d if d else m))
    return lin

for td in master_vips:
    if td.get("id") and td["id"] in individuals:
        full = list(reversed(climb(td["id"])))
        td["lin_str"] = " -> ".join([x["name"] for x in full])
        td["pids"] = ",".join([x["id"].replace('I','') for x in full])
        td["name"] = individuals[td["id"]]["name"]

rows = []
for vol in volume_list:
    uid = vol["match_id"]
    td = csv_auth[vol["tester_id"]]

    if uid not in individuals: continue
    p = individuals[uid]
    lin = climb(uid)
    if not lin: continue
    full = list(reversed(lin))

    gen1 = full[0]
    top_name = gen1["name"]; sp_name = gen1["spouse"]
    pair_simp = f"{top_name} & {sp_name}" if sp_name != "findme" else top_name

    b_yr = re.search(r'\d{4}', persons_data[gen1['id']].get('bdate', ''))
    d_yr = re.search(r'\d{4}', persons_data[gen1['id']].get('ddate', ''))
    dates = f"({b_yr.group(0) if b_yr else 'findme'} - {d_yr.group(0) if d_yr else 'findme'})"

    dir_lbl = f"{top_name} {dates}" + (f" & {sp_name}" if sp_name != "findme" else "")
    # 🚨 DOCTRINE FIX: Upstream Strict Normalization to prevent downstream UI/JS key mismatches
    dir_lbl = " ".join(dir_lbl.split())

    rows.append({
        "Tester_Code": td['id'].replace('I',''),
        "Tester_Name": td['name'],
        "Tester_ID": "",
        "Tester_Display": td['name'],
        "Tester_Sort_Key": td['sort_key'],
        "Tester_Lineage": td['lin_str'],
        "Tester_Path_IDs": td['pids'],
        "Match_Name": p["name"],
        "Match_ID": uid,
        "cM": vol["cm"],
        "Match_Lineage": " -> ".join([pair_simp] + [x["name"] for x in full[1:]]),
        "Match_Path_IDs": ",".join([x["id"].replace('I','') for x in full]),
        "Authority_Directory_Label": dir_lbl
    })

df = pd.DataFrame(rows)
if df.empty:
    print("    ⚠️ Notice: 0 matches processed.")
    db_json = []; participant_scores = []; anc_data = {}; derived_branches = {}; node_reviews = {}
else:
    df.to_csv(CSV_DB, index=False, encoding="iso-8859-15", quoting=csv.QUOTE_ALL)
    print(f"    ✅ Matrix Complete: {len(df)} lines written.")

    df.rename(columns={
        "Authority_Directory_Label": "Dir_Label",
        "Tester_Code": "Kit_Code",
        "Match_Lineage": "Lineage",
        "Match_Path_IDs": "s_ids",
        "Tester_Display": "Kit_Name"
    }, inplace=True)

    df['search_ids'] = df['s_ids']
    df['search_names'] = df['Lineage'].astype(str).str.replace(' -> ', '|')
    df['t_names'] = df['Tester_Lineage'].astype(str).str.replace(' -> ', '|')
    df['t_ids'] = df['Tester_Path_IDs'].astype(str).str.replace(',', '|')

    # --- STEP 5: MATHEMATICAL ENGINE (ANCHOR/CSS) ---
    print("    [+] Calculating CSS & DOCS Math Matrices...")
    def clean_num(s): return re.sub(r'[^0-9]', '', str(s))
    def norm_log(val, cap): return min(1.0, math.log(1 + max(0, val or 0)) / math.log(1 + cap))

    id_to_kits = {}
    for _, r in df.iterrows():
        for i in [clean_num(x) for x in str(r['s_ids']).split(',') if x]:
            id_to_kits.setdefault(i, set()).add(r['Kit_Name'])

    participant_scores = []
    for p_name, grp in df.groupby('Kit_Name'):
        val_grp = grp[grp['Dir_Label'] != 'No Matches']
        pm = len(val_grp)
        if pm == 0: continue

        dc = val_grp['Dir_Label'].value_counts()
        hc_t, hc_2 = (int(dc.iloc[0]), int(dc.iloc[1])) if len(dc)>1 else (int(dc.iloc[0]), 0) if len(dc)>0 else (0,0)
        target_anc = str(dc.index[0]) if len(dc)>0 else "Unknown"

        hh, target_id = 0, None
        for _, r in val_grp.iterrows():
            for i in [clean_num(x) for x in str(r['s_ids']).split(',') if x]:
                if len(id_to_kits.get(i, set())) > hh: hh, target_id = len(id_to_kits[i]), i

        spine_ids = []; br = 0; tb = 0
        dom_branch_pct = 0; branch_bal = "0 : 0"; dist_support = "Standard Convergence"

        if target_id:
            tb = len(id_to_kits.get(target_id, set()))
            col = df[df['s_ids'].apply(lambda x: target_id in [clean_num(y) for y in str(x).split(',') if y])]
            branch_counts = {}
            for _, r in col.iterrows():
                ids = [clean_num(x) for x in str(r['s_ids']).split(',') if x]
                nms = str(r['search_names']).split('|')
                if target_id in ids:
                    idx = ids.index(target_id)
                    b_name = nms[idx+1].replace('findme', '?').replace('FINDME', '?').split(' (')[0].strip() if idx+1 < len(nms) else "Direct"
                    branch_counts[b_name] = branch_counts.get(b_name, 0) + 1

            br = len(branch_counts)
            spine_ids = [clean_num(x) for x in str(val_grp[val_grp['s_ids'].apply(lambda x: target_id in str(x))].iloc[0]['s_ids']).split(',') if x]

            if branch_counts:
                sorted_counts = sorted(branch_counts.values(), reverse=True)
                max_b = sorted_counts[0]
                tot_b = sum(sorted_counts)
                if tot_b > 0:
                    dom_branch_pct = round((max_b / tot_b) * 100)
                    rest_pct = 100 - dom_branch_pct
                    branch_bal = f"{dom_branch_pct} : {rest_pct}"
                    if dom_branch_pct >= 80: dist_support = "Concentrated Lineage"
                    elif dom_branch_pct <= 50 and br >= 3: dist_support = "Highly Distributed"
                    else: dist_support = "Standard Convergence"

        dr = float(hc_t / max(1, hc_2))
        w_sum = norm_log(pm, 150) + norm_log(hc_t, 100) + (norm_log(dr, 10)*1.5) + norm_log(tb, 40) + norm_log(hh, 150)
        w_sum += 2.0 * (1.0 if br>=6 else 0.85 if br==5 else 0.70 if br==4 else 0.50 if br==3 else 0.25 if br==2 else 0)
        st_str, st_val = ("PASS", 1.0) if pm>=15 and br>=3 and dr>=1.5 else ("PARTIAL", 0.85) if pm>=15 and br>=2 else ("FAIL", 0.6)
        css = float(100 * (w_sum / 7.5) * st_val)

        diag_log = []
        for pid in spine_ids:
            d = persons_data.get(pid, {})
            if d.get('sources_count',0) + d.get('citations_count',0) == 0: diag_log.append(f"Missing Sources: {d.get('name')} (I{pid})")

        docs_score = 88.5 if len(diag_log)==0 else max(35.0, 85.0 - (len(diag_log)*5))
        anchor = min(100.0, css)

        participant_scores.append({
            'pName': str(p_name), 'targetAnc': target_anc, 'targetID': str(target_id or ''),
            'PM': int(pm), 'HC_T': int(hc_t), 'DR': float(dr), 'BR': int(br), 'NS': int(hh),
            'ST_str': st_str, 'cssFinal': float(css), 'DOCS': float(docs_score), 'ANCHOR': float(anchor),
            'diagLog': diag_log, 'AX': 1, 'CC': 1.0, 'BC': 1.0, 'DC': 1.0, 'TP': 1.0,
            'isGroup': False, 'GD': len(spine_ids),
            'domBranchPct': dom_branch_pct, 'branchBal': branch_bal, 'distSupport': dist_support
        })

    anc_data = {}
    derived_branches = {}
    node_reviews = {}

    print("    [+] Building Institutional Network Flow Topology (Generation by Generation)...")

    for lbl, grp in df.groupby('Dir_Label'):
        if str(lbl).strip() == 'No Matches' or not str(lbl).strip(): continue
        u_t = len(grp['Kit_Name'].unique())

        pure_pat = 0; pure_mat = 0; mixed = 0; unresolved = 0
        for _, r in grp.iterrows():
            ids = [clean_num(x) for x in str(r['s_ids']).split(',') if x]
            if len(ids) > 1:
                line_sexes = [individuals.get(f"I{iid}", {}).get('sex', 'U') for iid in ids[:-1]]
                if all(s == 'M' for s in line_sexes): pure_pat += 1
                elif all(s == 'F' for s in line_sexes): pure_mat += 1
                elif 'U' in line_sexes: unresolved += 1
                else: mixed += 1
            else:
                unresolved += 1

        anc_data[str(lbl)] = {
            "name": str(lbl), "matches": len(grp), "cm": int(grp['cM'].sum()),
            "badge": "Platinum" if len(grp)>=30 else "Gold",
            "integrity": min(100, (len(grp)*2) + (u_t*10)), "testers": u_t,
            "overall_case_confidence": "Moderate Confidence" if len(grp) >= 15 else "Low Confidence",
            "path_demographics": {
                "total_supporting": len(grp), "pure_matrilineal": pure_mat,
                "pure_patrilineal": pure_pat, "mixed_autosomal": mixed, "unresolved": unresolved
            }
        }

        branch_stats = {}; gen_stats = {}; total_matches = len(grp)

        for _, r in grp.iterrows():
            parts = str(r['Lineage']).split(' -> ')
            ids = [clean_num(x) for x in str(r['s_ids']).split(',') if x]

            # Gen 1 (Derived Branches)
            if len(parts) > 1: b_name, b_id = parts[1].strip(), (ids[1] if len(ids) > 1 else None)
            else: b_name, b_id = "Direct / Unknown", None

            if b_name not in branch_stats: branch_stats[b_name] = {'count': 0, 'id': b_id}
            branch_stats[b_name]['count'] += 1

            # Gen 2+ (Nodes)
            max_anc_idx = len(parts) - 1
            for g in range(2, max_anc_idx):
                p_name = parts[g-1].strip()
                node_name = parts[g].strip()
                node_id = ids[g] if len(ids) > g else None
                key = (p_name, node_name)

                if g not in gen_stats: gen_stats[g] = {}
                if key not in gen_stats[g]: gen_stats[g][key] = {'count': 0, 'id': node_id}
                gen_stats[g][key]['count'] += 1

        # Process Generation 1 Branches
        branch_list = []
        for b_name, stats in branch_stats.items():
            count = stats['count']
            pct = round((count / total_matches) * 100) if total_matches > 0 else 0

            contrib = "Dominant Contributor" if pct >= 60 else "Co-Dominant Contributor" if pct >= 30 else "Supporting Contributor" if pct >= 10 else "Marginal Contributor"

            b_id = stats['id']
            has_vitals = False; has_sources = False; sex_val = "U"; node_color = "#455a64"

            if b_id:
                clean_b_id = re.sub(r'[^0-9]', '', str(b_id))
                pd_obj = persons_data.get(clean_b_id) or persons_data.get("I" + clean_b_id) or {}
                if pd_obj.get('bdate') or pd_obj.get('ddate'): has_vitals = True
                if pd_obj.get('sources_count', 0) > 0 or (pd_obj.get('sources') and len(pd_obj.get('sources')) > 0): has_sources = True

                ind_obj = individuals.get(f"I{clean_b_id}")
                if ind_obj:
                    s = ind_obj.get('sex', 'U')
                    if s in ['M', 'F']:
                        sex_val = s
                        node_color = '#1565c0' if s == 'M' else '#ff00ff'

            supp = "Strong Support" if pct >= 40 and has_vitals and has_sources else "Adequate Support" if pct >= 15 and has_vitals else "Thin Support" if pct < 15 or not has_vitals else "Unresolved Support"

            notes = f"{pct}% of network flow."
            if not has_vitals: notes += " Missing core vitals."
            if not has_sources: notes += " Lacks direct sources."

            next_move = "Targeted documentary research." if supp == "Thin Support" else "Shift focus to Gen 2 nodes." if supp == "Strong Support" else "Expand collateral testing."

            branch_list.append({
                "derived_first_ancestor": b_name,
                "relationship": "Child",
                "contribution_pattern": contrib,
                "support_condition": supp,
                "support_notes": notes,
                "next_move": next_move,
                "count": count,
                "node_sex": sex_val,
                "node_id": f"I{clean_b_id}" if b_id else "",
                "node_color": node_color
            })

        branch_list.sort(key=lambda x: x['count'], reverse=True)
        derived_branches[str(lbl)] = branch_list
        active_b = len(branch_stats)
        anc_data[str(lbl)]["network_condition"] = f"Concentrated Lineage ({active_b} Active Branches)" if active_b <= 3 else f"Distributed Lineage ({active_b} Active Branches)"

        # Process Generation 2+ Nodes
        node_reviews_list = []
        for g, nodes in gen_stats.items():
            parent_totals = {}; parent_max = {}
            for (p_name, b_name), stats in nodes.items():
                c = stats['count']
                parent_totals[p_name] = parent_totals.get(p_name, 0) + c
                if c > parent_max.get(p_name, 0): parent_max[p_name] = c

            for (p_name, b_name), stats in nodes.items():
                count = stats['count']
                pct = round((count / total_matches) * 100) if total_matches > 0 else 0

                contrib = "Dominant Contributor" if pct >= 60 else "Co-Dominant Contributor" if pct >= 30 else "Supporting Contributor" if pct >= 10 else "Marginal Contributor"
                rel = "Primary Descending Line" if count == parent_max.get(p_name, 0) and count > 0 else "Collateral Sub-branch"

                b_id = stats['id']
                has_vitals = False; has_sources = False; sex_val = "U"; node_color = "#455a64"

                if b_id:
                    clean_b_id = re.sub(r'[^0-9]', '', str(b_id))
                    pd_obj = persons_data.get(clean_b_id) or persons_data.get("I" + clean_b_id) or {}
                    if pd_obj.get('bdate') or pd_obj.get('ddate'): has_vitals = True
                    if pd_obj.get('sources_count', 0) > 0 or (pd_obj.get('sources') and len(pd_obj.get('sources')) > 0): has_sources = True

                    ind_obj = individuals.get(f"I{clean_b_id}")
                    if ind_obj:
                        s = ind_obj.get('sex', 'U')
                        if s in ['M', 'F']:
                            sex_val = s
                            node_color = '#1565c0' if s == 'M' else '#ff00ff'

                supp = "Strong Support" if pct >= 40 and has_vitals and has_sources else "Adequate Support" if pct >= 15 and has_vitals else "Thin Support" if pct < 15 or not has_vitals else "Unresolved Support"

                notes = f"{pct}% of network flow."
                if not has_vitals: notes += " Missing core vitals."
                if not has_sources: notes += " Lacks direct sources."

                next_move = "Targeted documentary research." if supp == "Thin Support" else f"Shift focus to Gen {g+1} nodes." if supp == "Strong Support" else "Expand collateral testing."

                node_reviews_list.append({
                    "generation": g,
                    "node_ancestor": b_name,
                    "lead_emerged_ancestor": str(lbl),
                    "parent_line": p_name,
                    "relationship": rel,
                    "contribution_pattern": contrib,
                    "support_condition": supp,
                    "support_notes": notes,
                    "next_move": next_move,
                    "count": count,
                    "node_sex": sex_val,
                    "node_id": f"I{clean_b_id}" if b_id else "",
                    "node_color": node_color
                })

        node_reviews_list.sort(key=lambda x: (x['generation'], x['parent_line'], -x['count']))
        node_reviews[str(lbl)] = node_reviews_list

    export_cols = ['Dir_Label', 'Kit_Name', 'cM', 'Match_ID', 'Lineage', 's_ids', 'search_names', 't_names', 't_ids', 'Tester_ID', 'Kit_Code']
    rename_dict = {'Dir_Label': 'ancestor', 'Kit_Name': 'participant', 'cM': 'cm', 'Match_ID': 'id', 'Lineage': 'lineage', 'Tester_ID': 'tester_id', 'Kit_Code': 'kit_code'}
    db_json = df[export_cols].rename(columns=rename_dict).to_dict(orient='records')

    print("    ✅ Phase 3 Complete. Network Topology embedded directly into engine core.")
```

```python
# @title 🚀 [CELL 5B] The Topology, Cloud Media, & DDM Engine
def build_database_engine():
    print("="*60)
    print("      [CELL 5B] TOPOLOGY ENGINE STARTING (Cloud Sync & DDM Mode)...")
    print("="*60)

    import os, json, re, stat
    import pandas as pd

    if 'sftp' not in globals() or not transport.is_active():
        return print("\n❌ STOPPING: Environment not initialized. Run Cell 1 first.")

    JSON_DB = "compiled_database.json"
    CSV_DB = "engine_database.csv"
    TARGET_DIR = "yatesville.net/vault_yates_v2/vault-public-facing"

    if not os.path.exists(JSON_DB): return print("❌ ERROR: compiled_database.json not found.")
    with open(JSON_DB, "r", encoding="utf-8") as f: master_data = json.load(f)

    print("    [+] Loading Vault CSVs and SoA Protocols...")
    df = pd.read_csv(CSV_DB, encoding="iso-8859-15").fillna('').replace('nan', '')
    if "Authority_Directory_Label" in df.columns: df.rename(columns={"Authority_Directory_Label": "Dir_Label"}, inplace=True)

    for col in ["Tester_Display", "Kit_Name", "Participant_Name"]:
        if col in df.columns:
            df.rename(columns={col: "Participant_Name"}, inplace=True)
            df['Participant_Name'] = df['Participant_Name'].apply(lambda x: str(x).strip())
            break
    for col in ["Tester_Code", "Kit_Code", "Participant_Code"]:
        if col in df.columns:
            df.rename(columns={col: "Participant_Code"}, inplace=True)
            df['Participant_Code'] = df['Participant_Code'].apply(lambda x: str(x).strip())
            break
    for col in ["Match_Lineage", "Lineage"]:
        if col in df.columns: df.rename(columns={col: "Lineage"}, inplace=True); break

    df_valid = df[df['Dir_Label'] != 'No Matches'].copy()

    # --- SOURCE OF AUTHORITY SYNC ---
    auth_df = pd.DataFrame()
    for fname in ["key_participants.csv", "match_to_unmasked.csv"]:
        if os.path.exists(fname) and os.path.getsize(fname) > 0:
            try: auth_df = pd.read_csv(fname, encoding='utf-8-sig'); auth_df.columns = auth_df.columns.str.strip(); break
            except: continue

    id_to_sk, name_to_sk, sort_to_auth = {}, {}, {}
    if not auth_df.empty:
        for _, row in auth_df.iterrows():
            sk = str(row.get('Sort_Key_alpha', '')).strip()
            id_k = str(row.get('Id_Key', '')).strip().upper()
            code_k = str(row.get('Numeric_assigned', '')).strip()
            if sk:
                sort_to_auth[sk.lower()] = { 'id': id_k, 'code': code_k, 'sk_real': sk }
                name_to_sk[sk.lower()] = sk
            if id_k and sk: id_to_sk[id_k] = sk

    admin_rows = []
    part_dict = master_data.get('DATA', {}).get('participants', {})

    for p_name, p_info in part_dict.items():
        pure_name = str(p_info.get('name', p_name)).strip()
        kit_c = str(p_info.get('kit_code', '')).strip().upper()
        sk = id_to_sk.get(kit_c, str(p_info.get('sort_key', '')).strip())
        count = int(p_info.get('matches', 0))

        item = { 'name': pure_name, 'id': sort_to_auth.get(sk.lower(), {}).get('id', kit_c), 'count': count, 'sort': sk }
        admin_rows.append(item)

    admin_rows.sort(key=lambda x: x['sort'])

    soa_map = {r['name']: r['sort'].lower() for r in admin_rows}
    soa_map.update({r['id']: r['sort'].lower() for r in admin_rows})
    soa_map.update({r['id'].lower(): r['sort'].lower() for r in admin_rows})

    def get_soa_sk(p_name, p_code=""):
        if p_name in soa_map: return soa_map[p_name]
        p_c = str(p_code).strip().upper()
        if p_c in id_to_sk: return id_to_sk[p_c]
        n_low = str(p_name).strip().lower()
        if n_low in name_to_sk: return name_to_sk[n_low]
        for sk_low, real_sk in name_to_sk.items():
            if real_sk.lower() in n_low: return real_sk
        return str(p_name).strip()

    p_code_col = "Participant_Code" if "Participant_Code" in df_valid.columns else "Kit_Code"
    df_valid['sort_key'] = df_valid.apply(lambda x: get_soa_sk(x.get('Participant_Name', ''), x.get(p_code_col, '')), axis=1)

    def get_disp_name(sk):
        for r in admin_rows:
            if r['sort'].lower() == str(sk).lower(): return r['name']
        return ""
    df_valid['Participant_Name'] = df_valid['sort_key'].apply(lambda x: get_disp_name(x) or x)

    def clean_num(s): return re.sub(r'[^0-9]', '', str(s))

    # --- GEDCOM EXTRACTION & LEGACY DDM DATA ---
    print("    [+] Extracting GEDCOM Data (Vitals, Sources, Sex)...")
    gedcom_data = {}
    gedcom_sex_map = {}
    families = {}
    current_id = None; current_fam = None

    ged_files = [f for f in os.listdir('.') if f.lower().endswith('.ged')]
    if ged_files:
        with open(ged_files[0], 'r', encoding='utf-8', errors='replace') as gf:
            for line in gf:
                line = line.strip(); parts = line.split(" ", 2)
                if len(parts) < 2: continue
                lvl, tag = parts[0], parts[1]
                val = parts[2] if len(parts) > 2 else ""

                if lvl == "0" and "INDI" in val:
                    current_id = tag.replace("@", "")
                    gedcom_data[current_id] = {'sex': 'U', 'birth': False, 'death': False, 'marriage': False, 'place': False, 'source': False}
                    current_fam = None
                elif lvl == "0" and "FAM" in val:
                    current_fam = tag.replace("@", "")
                    families[current_fam] = {"husb": None, "wife": None, "marr": False}
                    current_id = None
                elif current_id and lvl != "0":
                    if tag == "SEX" and lvl == "1":
                        gedcom_data[current_id]['sex'] = val.upper()
                        gedcom_sex_map[current_id] = val.upper()
                    elif tag in ["BIRT", "CHR", "BAPM"]: gedcom_data[current_id]['birth'] = True
                    elif tag in ["DEAT", "BURI", "CREM"]: gedcom_data[current_id]['death'] = True
                    elif tag == "SOUR": gedcom_data[current_id]['source'] = True
                    elif tag == "PLAC" and val.strip(): gedcom_data[current_id]['place'] = True
                elif current_fam and lvl != "0":
                    if tag == "HUSB": families[current_fam]["husb"] = val.replace("@", "")
                    elif tag == "WIFE": families[current_fam]["wife"] = val.replace("@", "")
                    elif tag == "MARR": families[current_fam]["marr"] = True

        # Retroactively apply marriages
        for fam_id, fam_data in families.items():
            if fam_data.get("marr"):
                if fam_data["husb"] in gedcom_data: gedcom_data[fam_data["husb"]]['marriage'] = True
                if fam_data["wife"] in gedcom_data: gedcom_data[fam_data["wife"]]['marriage'] = True

    # --- TOPOLOGY REBUILDER ---
    print("    [+] Forcing Synchronized Data Matrix Rebuild...")
    persons_db = master_data.get('DATA', {}).get('persons', {})

    anc_data, derived_branches, node_reviews = {}, {}, {}

    for raw_lbl, grp in df_valid.groupby('Dir_Label'):
        lbl = " ".join(str(raw_lbl).split())
        if lbl == 'No Matches' or not lbl: continue

        total_matches = len(grp)
        anc_data[lbl] = {
            "name": lbl, "matches": total_matches,
            "cm": int(pd.to_numeric(grp['cM'].astype(str).str.replace(r'[^0-9.]', '', regex=True), errors='coerce').fillna(0).sum()),
            "testers": len(grp['Participant_Name'].unique()), "integrity": 100,
            "overall_case_confidence": "Moderate Confidence" if total_matches >= 15 else "Low Confidence"
        }

        branch_stats, gen_stats = {}, {}
        for _, r in grp.iterrows():
            parts = str(r['Lineage']).split(' -> ')
            ids = [clean_num(x) for x in str(r.get('s_ids', r.get('search_ids', ''))).split(',') if x]
            b_name, b_id = (parts[1].strip(), ids[1] if len(ids) > 1 else None) if len(parts) > 1 else ("Direct / Unknown", None)

            if b_name not in branch_stats: branch_stats[b_name] = {'count': 0, 'id': b_id}
            branch_stats[b_name]['count'] += 1

            for g in range(2, len(parts) - 1):
                p_name, node_name, node_id = parts[g-1].strip(), parts[g].strip(), ids[g] if len(ids) > g else None
                key = (p_name, node_name)
                if g not in gen_stats: gen_stats[g] = {}
                if key not in gen_stats[g]: gen_stats[g][key] = {'count': 0, 'id': node_id}
                gen_stats[g][key]['count'] += 1

        branch_list = []
        for b_name, stats in branch_stats.items():
            count, b_id = stats['count'], stats['id']
            pct = round((count / total_matches) * 100) if total_matches > 0 else 0
            contrib = "Dominant Contributor" if pct >= 60 else "Co-Dominant Contributor" if pct >= 30 else "Supporting Contributor" if pct >= 10 else "Marginal Contributor"
            has_vitals, has_sources, sex_val, sex_marker = False, False, "U", ""

            if b_id:
                clean_b_id = re.sub(r'[^0-9]', '', str(b_id))
                pd_obj = persons_db.get(clean_b_id) or persons_db.get("I" + clean_b_id) or {}
                has_vitals = bool(pd_obj.get('bdate') or pd_obj.get('ddate'))
                has_sources = bool(pd_obj.get('sources_count', 0) > 0 or pd_obj.get('sources'))
                sx = gedcom_sex_map.get(f"I{clean_b_id}", pd_obj.get('sex', pd_obj.get('gender', 'U')))
                if sx in ['M', 'F']: sex_val, sex_marker = sx, f" [{sx}]"

            supp = "Strong Support" if pct >= 40 and has_vitals and has_sources else "Adequate Support" if pct >= 15 and has_vitals else "Thin Support" if pct < 15 or not has_vitals else "Unresolved Support"
            notes = f"{pct}% of network flow. " + ("" if has_vitals else "Missing vitals. ") + ("" if has_sources else "No direct sources.")
            branch_list.append({ "derived_first_ancestor": f"{b_name}{sex_marker}", "relationship": "Child", "contribution_pattern": contrib, "support_condition": supp, "support_notes": notes.strip(), "next_move": "Targeted documentary research." if supp == "Thin Support" else "Shift focus to Gen 2.", "count": count, "node_sex": sex_val, "node_id": f"I{clean_b_id}" if b_id else "" })

        derived_branches[lbl] = sorted(branch_list, key=lambda x: x['count'], reverse=True)

        node_reviews_list = []
        for g, nodes in gen_stats.items():
            parent_max = {p: max((s['count'] for (pr, _), s in nodes.items() if pr == p), default=0) for (p, _) in nodes.keys()}
            for (p_name, b_name), stats in nodes.items():
                count, b_id = stats['count'], stats['id']
                pct = round((count / total_matches) * 100) if total_matches > 0 else 0
                contrib = "Dominant" if pct >= 60 else "Co-Dominant" if pct >= 30 else "Supporting" if pct >= 10 else "Marginal"
                rel = "Primary Descending Line" if count == parent_max.get(p_name, 0) and count > 0 else "Collateral Sub-branch"

                has_vitals, has_sources, sex_val = False, False, "U"
                if b_id:
                    clean_b_id = re.sub(r'[^0-9]', '', str(b_id))
                    pd_obj = persons_db.get(clean_b_id) or persons_db.get("I" + clean_b_id) or {}
                    has_vitals, has_sources = bool(pd_obj.get('bdate') or pd_obj.get('ddate')), bool(pd_obj.get('sources_count', 0) > 0 or pd_obj.get('sources'))
                    sx = gedcom_sex_map.get(f"I{clean_b_id}", pd_obj.get('sex', pd_obj.get('gender', 'U')))
                    if sx in ['M', 'F']: sex_val = sx

                supp = "Strong Support" if pct >= 40 and has_vitals and has_sources else "Adequate Support" if pct >= 15 and has_vitals else "Thin Support" if pct < 15 or not has_vitals else "Unresolved Support"
                node_reviews_list.append({ "generation": g, "node_ancestor": b_name, "parent_line": p_name, "relationship": rel, "contribution_pattern": f"{contrib} Contributor", "support_condition": supp, "support_notes": f"{pct}% flow. " + ("" if has_vitals else "Missing vitals. ") + ("" if has_sources else "No sources."), "next_move": "Targeted research." if supp == "Thin Support" else f"Shift to Gen {g+1}.", "count": count, "node_sex": sex_val, "node_id": f"I{clean_b_id}" if b_id else "" })

        node_reviews[lbl] = sorted(node_reviews_list, key=lambda x: (x['generation'], x['parent_line'], -x['count']))

    if 'DATA' not in master_data: master_data['DATA'] = {}
    master_data['DATA']['ancestors'], master_data['DATA']['derived_branches'], master_data['DATA']['node_reviews'] = anc_data, derived_branches, node_reviews

    def get_node_color(item):
        node_id = item.get('node_id')
        if not node_id: return '#455a64'
        node_id = f"I{node_id.replace('I','')}" if not node_id.startswith('I') else node_id
        sx = gedcom_sex_map.get(node_id, persons_db.get(node_id, {}).get('sex', persons_db.get(node_id, {}).get('gender', ''))).upper()
        return '#1565c0' if sx == 'M' else '#ff00ff' if sx == 'F' else '#455a64'

    for arr in list(derived_branches.values()) + list(node_reviews.values()):
        for item in arr: item['node_color'] = get_node_color(item)

    precomputed = master_data.get('PRECOMPUTED', [])
    for p in precomputed:
        p_name = p.get('pName', '')
        p['kit_id'] = next((r['id'] for r in admin_rows if r['name'] == p_name), "")
        p['sortKey'] = get_soa_sk(p_name)

    # --- 🚨 LEGACY DDM MATH INJECTION ---
    print("    [+] Calculating Legacy Normalized DDM Scores...")
    part_spines = {}
    for p_name, grp in df_valid.groupby('Participant_Name'):
        raw_spine = []
        for _, r in grp.iterrows():
            raw_spine.extend([clean_num(x) for x in str(r.get('s_ids', r.get('search_ids', ''))).split(',') if x])
        part_spines[p_name] = raw_spine

    def get_gedcom_pid(pid):
        c = clean_num(pid)
        if f"I{c}" in gedcom_data: return f"I{c}"
        if c in gedcom_data: return c
        return None

    def calc_docs_rate(spines):
        valid_pids = [get_gedcom_pid(x) for x in spines if get_gedcom_pid(x)]
        unique_spines = set(valid_pids)
        b_t = d_t = m_t = p_t = s_t = 0
        for pid in unique_spines:
            dbin = gedcom_data[pid]
            if dbin['birth']: b_t += 1
            if dbin['death']: d_t += 1
            if dbin['marriage']: m_t += 1
            if dbin['place']: p_t += 1
            if dbin['source']: s_t += 1
        s_len = len(unique_spines)
        if s_len == 0: return {'b':0, 'd':0, 'm':0, 'p':0, 's':0, 'len':0, 'raw_b':0, 'raw_d':0, 'raw_m':0, 'raw_p':0, 'raw_s':0}
        return {
            'b': int(round((b_t/s_len)*100)), 'd': int(round((d_t/s_len)*100)),
            'm': int(round((m_t/s_len)*100)), 'p': int(round((p_t/s_len)*100)),
            's': int(round((s_t/s_len)*100)), 'len': s_len,
            'raw_b': b_t, 'raw_d': d_t, 'raw_m': m_t, 'raw_p': p_t, 'raw_s': s_t
        }

    # Map to precomputed
    for p in precomputed:
        p_name = p.get('pName', '')
        p['docs_rates'] = calc_docs_rate(part_spines.get(p_name, []))

    # Map to ancestors
    for lbl, data in anc_data.items():
        grp = df_valid[df_valid['Dir_Label'] == lbl]
        raw_spine = []
        for _, r in grp.iterrows():
            raw_spine.extend([clean_num(x) for x in str(r.get('s_ids', r.get('search_ids', ''))).split(',') if x])
        data['docs_rates'] = calc_docs_rate(raw_spine)

    def apply_quartiles(data_list, dict_key):
        keys = ['b', 'd', 'm', 'p', 's']
        for k in keys:
            rates = [item[dict_key][k] for item in data_list if dict_key in item and item[dict_key].get('len', 0) > 0]
            q2 = pd.Series(rates).quantile(0.50) if rates else 0
            q3 = pd.Series(rates).quantile(0.75) if rates else 0
            for item in data_list:
                if dict_key not in item: continue
                d = item[dict_key]
                if d.get('len', 0) == 0:
                    d[f"{k}_str"] = "— ⚪"; d[f"{k}_tooltip"] = "No spine calculated"
                else:
                    val = d[k]; raw_val = d.get(f"raw_{k}", 0)
                    if val == 0: color = "⚫"
                    elif val >= q3 and val > 0: color = "🟢"
                    elif val >= q2 and val > 0: color = "🟡"
                    else: color = "🔴"
                    d[f"{k}_str"] = f"{val} {color}"; d[f"{k}_tooltip"] = f"{raw_val} of {d['len']} persons"

    apply_quartiles(precomputed, 'docs_rates')
    apply_quartiles(list(anc_data.values()), 'docs_rates')

    # --- CLOUD MEDIA SCAFFOLDING & SCANNER ---
    print("    [+] Engaging Cloud SFTP Crawler (DreamHost Media Vault)...")
    remote_media_dir = f"{TARGET_DIR}/Media"

    try: sftp.stat(remote_media_dir)
    except IOError: sftp.mkdir(remote_media_dir)

    try: sftp.stat(f"{remote_media_dir}/all")
    except IOError: sftp.mkdir(f"{remote_media_dir}/all")

    created_dirs = 0
    required_nodes = set()

    # Gather all required node IDs from participants
    for p in admin_rows:
        c_num = clean_num(str(p['id']))
        if c_num: required_nodes.add(f"I{c_num}")

    # Gather all required node IDs from topology ancestors
    for raw_lbl, grp in df_valid.groupby('Dir_Label'):
        s_ids_str = str(grp.get('s_ids', pd.Series([''])).iloc[0]) if 's_ids' in grp.columns else str(grp.get('search_ids', pd.Series([''])).iloc[0])
        for x in s_ids_str.split(','):
            c_num = clean_num(x)
            if c_num: required_nodes.add(f"I{c_num}")

    # List existing DreamHost directories to minimize SFTP commands
    existing_remote_dirs = []
    try:
        for item in sftp.listdir_attr(remote_media_dir):
            if stat.S_ISDIR(item.st_mode):
                existing_remote_dirs.append(item.filename)
    except IOError:
        pass

    for node in required_nodes:
        if node not in existing_remote_dirs:
            try:
                sftp.mkdir(f"{remote_media_dir}/{node}")
                created_dirs += 1
            except IOError: pass

    if created_dirs > 0:
        print(f"        -> Cloud Scaffolding verified. Created {created_dirs} new 'I-Node' folders on DreamHost.")
    else:
        print(f"        -> Cloud Scaffolding verified. All {len(required_nodes)} folders already exist on DreamHost.")

    print("    [+] Scanning DreamHost for Active Media Files...")
    all_media = []
    id_media_map = {}

    dirs_to_explore = [remote_media_dir]
    file_count = 0

    while dirs_to_explore:
        current_dir = dirs_to_explore.pop(0)
        try:
            for item in sftp.listdir_attr(current_dir):
                if stat.S_ISDIR(item.st_mode):
                    dirs_to_explore.append(f"{current_dir}/{item.filename}")
                else:
                    if item.filename.lower().endswith(('.pdf', '.png', '.jpg', '.jpeg', '.mp4', '.mp3', '.mkv', '.wav')):
                        rel_path = current_dir.replace(remote_media_dir, "").strip("/")
                        if not rel_path: continue

                        folder_name = rel_path.split('/')[0].lower()
                        web_path = f"Media/{rel_path}/{item.filename}"
                        file_count += 1

                        if folder_name == 'all':
                            all_media.append(web_path)
                        elif folder_name.startswith('i'):
                            clean_i = folder_name.upper()
                            if clean_i not in id_media_map: id_media_map[clean_i] = []
                            id_media_map[clean_i].append(web_path)
        except IOError:
            pass

    print(f"        -> Discovered {file_count} valid media files across the Cloud Vault.")

    # 1. Map Media to Ancestors
    media_dict = {}
    for raw_lbl, grp in df_valid.groupby('Dir_Label'):
        lbl = " ".join(str(raw_lbl).split())
        if not lbl or lbl == 'No Matches': continue
        s_ids_str = str(grp.get('s_ids', pd.Series([''])).iloc[0]) if 's_ids' in grp.columns else str(grp.get('search_ids', pd.Series([''])).iloc[0])
        ids = [clean_num(x) for x in s_ids_str.split(',') if x]

        lbl_media = list(all_media)
        for i_num in ids[:3]:
            if f"I{i_num}" in id_media_map: lbl_media.extend(id_media_map[f"I{i_num}"])
        if lbl_media: media_dict[lbl] = list(set(lbl_media))

    # 2. Map Media to Participants
    for p in admin_rows:
        p_name = p['name']
        c_num = clean_num(str(p['id']))
        if c_num:
            p_id = f"I{c_num}"
            p_media = list(all_media)
            if p_id in id_media_map: p_media.extend(id_media_map[p_id])
            if p_media: media_dict[p_name] = list(set(p_media))

    master_data['MEDIA'] = media_dict

    with open(JSON_DB, "w", encoding="utf-8") as f: json.dump(master_data, f)

    globals()['telemetry_active_testers'] = len(df_valid['Participant_Name'].unique()) if not df_valid.empty else 0
    globals()['telemetry_total_matches'] = len(df_valid)
    print("✅ CLOUD & DDM SYNTHESIS COMPLETE. Topology, Media & Math saved to memory.")
    print("➡️ PROCEED TO CELL 5C TO INJECT UI.")

build_database_engine()
```

```python
# @title 🗄️ [CELL 7] The Time Machine (v2 Vault Archive & Sync)
def run_archiver():
    print("="*60)
    print("      🗄️ [CELL 7] THE TIME MACHINE (v2 Vault Archive & Sync)")
    print("="*60)

    import zipfile, os, pytz, json
    from datetime import datetime
    from google.colab import files
    try: from google.colab import userdata
    except ImportError: pass

    if 'sftp' not in globals():
        return print("❌ ERROR: SFTP pipe is not open. Please run Cell 1 first.")

    est = pytz.timezone('US/Eastern')
    timestamp = datetime.now(est).strftime("%Y-%m-%d_%H%M_%S")
    target_dir = 'site' if os.path.exists('site') else '.'

    # 1. GENERATE LEGACY README
    print("    [STEP 1] Generating System Handover Protocol...")
    readme_content = f"""=========================================================
YATES ONS STUDY - SYSTEM ENGINE & DATA ARCHIVE
Archive Date: {datetime.now(est).strftime("%B %d, %Y")}
=========================================================
TO THE FUTURE RESEARCHER:
You have inherited the complete, self-sustaining dataset and processing engine
for the Yates DNA Study. Everything required to rebuild and publish this
genealogical network is contained within this folder.
========================================================="""

    readme_name = "000_START_HERE_YATES_SYSTEM.txt"
    with open(readme_name, "w", encoding="utf-8") as f: f.write(readme_content)

    # 2. GENERATE SITE SNAPSHOT JSON
    print("    [STEP 2] Generating Site Snapshot JSON...")
    snapshot_data = {}
    html_files = [f for f in os.listdir(target_dir) if f.lower().endswith(('.shtml', '.html'))]
    for f_name in html_files:
        try:
            with open(os.path.join(target_dir, f_name), 'r', encoding='utf-8') as fh:
                snapshot_data[f_name] = fh.read()
        except: pass

    snapshot_name = f"site_snapshot_{timestamp}.json"
    with open(snapshot_name, 'w', encoding='utf-8') as f: json.dump(snapshot_data, f)

    # 3. CREATE MASTER ZIP VAULT
    extensions = ('.csv', '.shtml', '.html', '.json', '.js', '.css', '.ged', '.ipynb', '.txt')
    files_to_pack = [readme_name]

    for root_dir, dirs, files_in_dir in os.walk('.'):
        if '.config' in root_dir or 'sample_data' in root_dir: continue
        for f in files_in_dir:
            if f.lower().endswith(extensions) and f != snapshot_name:
                files_to_pack.append(os.path.join(root_dir, f))

    zip_name = f"Yates_Study_v2_Vault_{timestamp}.zip"
    print(f"    [STEP 3] Compressing {len(files_to_pack)} files into {zip_name}...")
    try:
        with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zf:
            for file in files_to_pack:
                arcname = file.replace('./', '') if file.startswith('./') else file
                zf.write(file, arcname)
        print(f"    ✅ Archive Created: {os.path.getsize(zip_name)/1024/1024:.2f} MB")
    except Exception as e:
        return print(f"    ❌ Compression Failed: {e}")

    # 4. SFTP UPLOAD TO VAULT ARCHIVE & ROLLING 10 CLEANUP
    print("\n    [STEP 4] Pushing to v2 Memory Archive via SFTP...")

    # 🛑 PATH LOCK: Updated to v2 Architecture -> vault_admin/memory
    ARCHIVE_DIR = "yatesville.net/vault_yates_v2/vault_admin/memory"

    try:
        sftp.put(zip_name, f"{ARCHIVE_DIR}/{zip_name}")
        print(f"    ✅ Uploaded to: {ARCHIVE_DIR}/{zip_name}")

        print("    [*] Running Vault Cleanup (Rolling 10 Policy)...")
        server_files = sftp.listdir(ARCHIVE_DIR)
        zip_backups = sorted([f for f in server_files if f.startswith("Yates_Study_v2_Vault_") and f.endswith(".zip")])

        zips_to_delete = zip_backups[:-10]
        if not zips_to_delete: print("      -> Archive is under 10 backups. Clean.")
        else:
            for old_zip in zips_to_delete:
                sftp.remove(f"{ARCHIVE_DIR}/{old_zip}")
                print(f"      🗑️ Incinerated old Vault ZIP: {old_zip}")
    except Exception as e: print(f"    ⚠️ Vault Archive Upload failed: {e}")

    # 5. DROPBOX SYNC
    print("\n    [STEP 5] Syncing to Dropbox Omni-Vault...")
    try:
        dbx_app_key = os.environ.get("DBX_APP_KEY") or userdata.get("DBX_APP_KEY")
        dbx_app_secret = os.environ.get("DBX_APP_SECRET") or userdata.get("DBX_APP_SECRET")
        dbx_refresh = os.environ.get("DBX_REFRESH_TOKEN") or userdata.get("DBX_REFRESH_TOKEN")

        if not dbx_refresh: print("    ❌ 'DBX_REFRESH_TOKEN' not found.")
        else:
            try: import dropbox
            except ImportError: os.system('pip install dropbox -q'); import dropbox

            dbx = dropbox.Dropbox(app_key=dbx_app_key, app_secret=dbx_app_secret, oauth2_refresh_token=dbx_refresh)
            folder_path = "/Yates_Study_Sync/archives"
            zip_target = f"{folder_path}/{zip_name}"

            with open(zip_name, "rb") as f: dbx.files_upload(f.read(), zip_target)
            print(f"    ✅ Dropbox Sync Success")

            # Dropbox Sweep
            result = dbx.files_list_folder(folder_path)
            dbx_files = [entry.name for entry in result.entries if isinstance(entry, dropbox.files.FileMetadata)]
            dbx_zips = sorted([f for f in dbx_files if f.startswith("Yates_Study_v2_Vault_") and f.endswith(".zip")])
            for old_zip in dbx_zips[:-10]:
                dbx.files_delete_v2(f"{folder_path}/{old_zip}")
    except Exception as e: print(f"    ❌ Dropbox Upload Failed: {e}")

    print("\n    [STEP 6] Triggering Local Download...")
    try: files.download(zip_name)
    except: pass

run_archiver()
```

```python
# @title 🔍 [CELL 8] The Standalone Deployment Crawler (Live Output Sync)
def run_standalone_crawler():
    print("="*60)
    print("      [CELL 8] DEPLOYMENT CRAWLER STARTING (Independent Run)...")
    print("="*60)

    import os, re
    from datetime import datetime
    import pytz

    if 'sftp' not in globals() or not transport.is_active():
        return print("❌ STOPPING: SFTP pipe closed. Run Cell 1 first.")

    est = pytz.timezone('US/Eastern')
    TARGET_DIR = "yatesville.net/vault_yates_v2/vault-public-facing"
    V2_ROOT = "https://yatesville.net/vault_yates_v2/vault-public-facing"

    print(f"[*] Scanning remote deployment directory: {TARGET_DIR}")

    try:
        attrs = sftp.listdir_attr(TARGET_DIR)
        web_files = [f for f in attrs if f.filename.endswith(('.html', '.shtml', '.htm'))]

        print(f"    ✅ Found {len(web_files)} web files deployed on the server.\n")

        if not os.path.exists('site'): os.makedirs('site', exist_ok=True)

        # Build a dictionary of actual files on server for fast lookup
        server_files_mtime = {}
        for f in web_files:
            server_files_mtime[f.filename] = datetime.fromtimestamp(f.st_mtime, est).strftime('%Y-%m-%d %I:%M:%S %p')

        # =====================================================================
        # 1. EXTRACT GLOBAL NAVIGATION MAP
        # =====================================================================
        print("    [+] Extracting Global Navigation link mapping...")
        remote_nav = f"{TARGET_DIR}/global_nav.html"
        local_nav = "site/global_nav_temp.html"
        nav_items = []

        try:
            sftp.get(remote_nav, local_nav)
            with open(local_nav, "r", encoding="utf-8") as f:
                nav_content = f.read()

            # Find all <a href="...">Text</a> tags
            matches = re.findall(r'<a[^>]+href=["\']([^"\']+)["\'][^>]*>(.*?)</a>', nav_content, re.IGNORECASE)
            for href, text in matches:
                clean_text = re.sub(r'<[^>]+>', '', text).strip()
                clean_filename = href.split('?')[0].split('/')[-1]
                full_url = href if href.startswith('http') else f"{V2_ROOT}/{clean_filename}"
                nav_items.append((clean_text, clean_filename, full_url))

            if os.path.exists(local_nav): os.remove(local_nav)
        except Exception as e:
            print(f"    ⚠️ Could not parse global_nav.html (it may not exist yet): {e}")

        # =====================================================================
        # 2. DEFINE THE INTENDED MACRO-SUITE (Pruned to match Cell 5C)
        # =====================================================================
        expected_macro_suite = {
            "about.shtml": "About",
            "anchor.shtml": "ANCHOR Database",
            "reports_exhibits.shtml": "Reports & Exhibits",
            "participant_profile.html": "Participant Profile",
            "dna_network.html": "Network Visualizer",
            "research_admin.html": "Admin Hub",
            "proof_consolidator.html": "Proof Consolidator",
            "zero_matches.html": "Zero Matches Review",
            "just-trees.shtml": "Yates DNA Trees",
            "ons_yates_dna_register.shtml": "ONS Yates DNA Register"
        }

        # =====================================================================
        # 3. DEFINE HIERARCHICAL RAW FILE CATEGORIES (For remaining audit)
        # =====================================================================
        core_files = ['anchor.shtml', 'participant_profile.html', 'research_admin.html', 'global_nav.html']
        sub_files = list(expected_macro_suite.keys())

        group_core, group_sub, group_support = [], [], []

        for f in web_files:
            mod_time = server_files_mtime[f.filename]
            live_url = f"{V2_ROOT}/{f.filename}"

            if f.filename in core_files: group_core.append((f.filename, mod_time, live_url))
            elif f.filename in sub_files: group_sub.append((f.filename, mod_time, live_url))
            else: group_support.append((f.filename, mod_time, live_url))

        group_core.sort(key=lambda x: core_files.index(x[0]))
        group_sub.sort(key=lambda x: sub_files.index(x[0]) if x[0] in sub_files else len(sub_files))
        group_support.sort(key=lambda x: x[0])

        def build_row(item):
            fn, mt, url = item
            icon = "📄"
            if fn == 'global_nav.html': icon = "🌐"
            elif fn == 'anchor.shtml': icon = "⚓"
            elif fn == 'research_admin.html': icon = "⚙️"
            elif fn == 'participant_profile.html': icon = "👥"
            return f"<tr><td style='text-align:left;'>{icon} <strong>{fn}</strong></td><td>{mt}</td><td style='text-align:left;'><a href='{url}' target='_blank' style='color:#1d4f91; text-decoration:none;'>{url}</a></td></tr>"

        # =====================================================================
        # 4. ASSEMBLE HTML ROWS
        # =====================================================================
        live_rows = []

        # A. Navigation Map
        if nav_items:
            live_rows.append("<tr style='background:#1d4f91; color:white;'><td colspan='3' style='text-align:left; font-weight:bold; font-size:15px; letter-spacing:1px;'>🌐 I. CURRENT GLOBAL NAVIGATION MAP</td></tr>")
            for text, fname, url in nav_items:
                live_rows.append(f"<tr style='background:#f8fafc;'><td style='text-align:left; padding-left:25px;'>🔗 <strong>{text}</strong></td><td><span style='color:#64748b; font-size:13px;'>Target:</span> {fname}</td><td style='text-align:left;'><a href='{url}' target='_blank' style='color:#1d4f91; text-decoration:none;'>{url}</a></td></tr>")

        # B. Macro-Suite Intended Deployment Map
        live_rows.append("<tr style='background:#15803d; color:white;'><td colspan='3' style='text-align:left; font-weight:bold; font-size:15px; letter-spacing:1px;'>📦 II. MACRO-SUITE DEPLOYMENT MAP (pages_to_upload)</td></tr>")
        for fname, title in expected_macro_suite.items():
            url = f"{V2_ROOT}/{fname}"
            # Check if file is actually on server
            mt = server_files_mtime.get(fname, "<span style='color:#dc2626; font-weight:bold;'>⚠️ MISSING FROM SERVER</span>")
            live_rows.append(f"<tr style='background:#f0fdf4;'><td style='text-align:left; padding-left:25px;'>📦 <strong>{fname}</strong></td><td>{mt}</td><td style='text-align:left;'><a href='{url}' target='_blank' style='color:#15803d; text-decoration:none;'>{url}</a></td></tr>")

        # C. Raw Directory Audit
        live_rows.append("<tr style='background:#334155; color:white;'><td colspan='3' style='text-align:left; font-weight:bold; font-size:15px; letter-spacing:1px;'>🔍 III. RAW SERVER DIRECTORY AUDIT</td></tr>")
        if group_core:
            live_rows.append("<tr style='background:#e2e8f0;'><td colspan='3' style='text-align:left; font-weight:bold; color:#334155; font-size:14px; padding-left:25px;'>Core Hubs & Nav</td></tr>")
            for item in group_core: live_rows.append(build_row(item))

        if group_support:
            live_rows.append("<tr style='background:#e2e8f0;'><td colspan='3' style='text-align:left; font-weight:bold; color:#334155; font-size:14px; padding-left:25px;'>Orphaned, Archived, or Support Pages (Not in Macro-Suite)</td></tr>")
            for item in group_support: live_rows.append(build_row(item))

        live_rows_html = "".join(live_rows)

        # =====================================================================
        # 5. SURGICAL INJECTION INTO ADMIN HUB
        # =====================================================================
        print("    [+] Downloading live Admin Hub for surgical injection...")
        remote_admin = f"{TARGET_DIR}/research_admin.html"
        local_admin = "site/research_admin_temp.html"

        try:
            # Download the live file
            sftp.get(remote_admin, local_admin)
            with open(local_admin, "r", encoding="utf-8") as f:
                admin_content = f.read()

            # Surgically replace only the table body
            new_admin_content = re.sub(
                r'<tbody id="live-output-body">.*?</tbody>',
                f'<tbody id="live-output-body">{live_rows_html}</tbody>',
                admin_content,
                flags=re.DOTALL
            )

            # Save and upload back
            with open(local_admin, "w", encoding="utf-8") as f:
                f.write(new_admin_content)

            sftp.put(local_admin, remote_admin)
            if os.path.exists(local_admin): os.remove(local_admin)

            print("    ✅ SUCCESS: Live Page Output sorted, pruned, and synced to Admin Hub.")
            print(f"👉 You can now review the Live URLs at: {V2_ROOT}/research_admin.html")

        except Exception as e:
            print(f"    ❌ Failed to update Admin Hub. Ensure Cell 5C has run at least once to create research_admin.html: {e}")

    except Exception as e:
        print(f"❌ ERROR during SFTP crawl: {e}")

run_standalone_crawler()
```
