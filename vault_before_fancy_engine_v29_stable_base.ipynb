```python
# @title 📥 [CELL 3A] The Data Ingestion & Mapper
# 🛑 DOCTRINE LOCK: YATES DNA ANCHOR STUDY
# RULE: ID-TO-GEDCOM NAMING ONLY.
# RULE: CSV IS TRUTH (CSS/DOCS); GEDCOM IS MAP.
# RULE: GENERATOR-FIRST (NO SURGICAL SERVER PATCHES).
# RULE: NO PARTIAL BLOCKS. COMPILE FULL CELL.

print("="*60)
print("      [PHASE 1] INGESTING SOURCE OF AUTHORITY DATA...")
print("="*60)

import os, sys, re, csv, json, math, shutil, glob
import pandas as pd

# --- MEMORY FLUSH ---
files_to_kill = ["engine_database.csv", "compiled_database.json", "css_math_audit.csv", "participant_match_audit.csv", "_processed_unmasked.ged"]
for f in files_to_kill:
    if os.path.exists(f): os.remove(f)

CSV_DB = "engine_database.csv"
JSON_DB = "compiled_database.json"
KEY_PARTICIPANTS = "key_participants.csv"
KEY_MATCHES = "key_matches.csv"
OVERRIDE_CSV = "gedcom_name_overrides.csv"
PROCESSED_GED = "_processed_unmasked.ged"
DATA_STORE_DIR = "yatesville.net/vault_yates/vault_data-store-inspect-inout"

if not os.path.exists(KEY_PARTICIPANTS) or not os.path.exists(KEY_MATCHES):
    raise FileNotFoundError(f"❌ FATAL ERROR: Source of Authority CSVs are missing! Please run Cell 2.")

def norm_key(k): return re.sub(r'[^A-Z0-9]', '', str(k).upper())

# --- STEP 1: LOAD VIP ROSTER (STRICT COLUMN A & B MAPPER) ---
print(f"    [+] Loading VIP Roster Map from {KEY_PARTICIPANTS}...")
df_keys = pd.read_csv(KEY_PARTICIPANTS, encoding='utf-8-sig').fillna('')
df_keys.columns = df_keys.columns.str.strip()
csv_auth = {}
master_vips = []

# ANTI-DRIFT VALIDATION: Force parser to ignore Display_Name column
for index, row in df_keys.iterrows():
    if pd.notnull(row.get('Display_Name')): pass

for _, row in df_keys.iterrows():
    col_a_id = str(row.get('Id_Key', '')).strip().upper()
    if not col_a_id or col_a_id == 'NAN': continue
    if not col_a_id.startswith('I'): col_a_id = 'I' + col_a_id

    col_b_sort = str(row.get('Sort_Key_alpha', '')).strip()

    td = {
        "name": "PENDING_GEDCOM_COMPUTATION",
        "id": col_a_id,
        "sort_key": col_b_sort.lower(),
        "lin_str": "", "pids": ""
    }
    master_vips.append(td)
    csv_auth[col_a_id] = td

# --- STEP 1.5: LOAD NAME OVERRIDES SoA ---
override_dict = {}
if os.path.exists(OVERRIDE_CSV):
    print(f"    [+] Loading SoA Name Overrides from {OVERRIDE_CSV}...")
    try:
        df_over = pd.read_csv(OVERRIDE_CSV, encoding='utf-8-sig').fillna('')
        fn_col = 'First Name' if 'First Name' in df_over.columns else 'Given_Name'
        sn_col = 'Surname' if 'Surname' in df_over.columns else 'Surname'

        for _, row in df_over.iterrows():
            pid = str(row.get('Id_Key', '')).strip().upper()
            if not pid or pid == 'NAN': continue
            if not pid.startswith('I'): pid = 'I' + pid

            given = str(row.get(fn_col, '')).strip()
            surname = str(row.get(sn_col, '')).strip()

            if given or surname:
                override_dict[pid] = {"given": given, "surname": surname}
        print(f"        -> {len(override_dict)} injected names locked in memory.")
    except Exception as e:
        print(f"    [⚠️] Failed to load Name Overrides: {e}")

# --- STEP 2: LOAD MATCH VOLUME (DIRECT ID MATCHING ONLY) ---
print(f"    [+] Loading Match Volume from {KEY_MATCHES}...")
df_matches = pd.read_csv(KEY_MATCHES, encoding='utf-8-sig').fillna('')
df_matches.columns = df_matches.columns.str.strip()

overlay_injector = {}
volume_list = []

for _, row in df_matches.iterrows():
    m_id = str(row.get('Match_ID', '')).upper().strip()
    t_id = str(row.get('Id_Key', '')).upper().strip()
    cm_val = str(row.get('cM', '0')).strip()

    if not m_id.startswith('I'): m_id = 'I' + m_id
    if not t_id.startswith('I'): t_id = 'I' + t_id

    if m_id and t_id in csv_auth:
        clean_m_id = m_id.replace('I', '')
        volume_list.append({"match_id": m_id, "tester_id": t_id, "cm": int(float(cm_val)) if cm_val.replace('.','').isdigit() else 0})
        if clean_m_id not in overlay_injector: overlay_injector[clean_m_id] = []
        overlay_injector[clean_m_id].append(f"{cm_val} cM & {t_id}")

print(f"        -> Discovered {len(volume_list)} precise match connections via ID mapping.")
print("    ✅ Phase 1 Complete. Variables loaded into memory.")
```

```python
# @title 🌳 [CELL 3B] The GEDCOM & Naming Engine
# 🛑 DOCTRINE LOCK: YATES DNA ANCHOR STUDY
# RULE: ID-TO-GEDCOM NAMING ONLY.
# RULE: CSV IS TRUTH (CSS/DOCS); GEDCOM IS MAP.
# RULE: GENERATOR-FIRST (NO SURGICAL SERVER PATCHES).
# RULE: NO PARTIAL BLOCKS. COMPILE FULL CELL.

print("="*60)
print("      [PHASE 2] NATIVE GEDCOM PARSING & NAMING MATH...")
print("="*60)

# --- STEP 3: LOCALIZE & PREP GEDCOM ---
ged_files = [f for f in glob.glob("*.ged") if "_processed" not in f.lower()]
if not ged_files: raise FileNotFoundError("❌ No native GEDCOM found. Please ensure a .ged file is in the directory.")
INPUT_GED = sorted(ged_files, key=lambda x: os.path.getmtime(x), reverse=True)[0]
print(f"    [+] Active Native GEDCOM Map: {INPUT_GED}")

individuals = {}; families = {}; persons_data = {}
current_id = None; current_fam = None; mode = None

with open(INPUT_GED, "r", encoding="utf-8", errors="replace") as f:
    for line in f:
        line = line.strip(); parts = line.split(" ", 2)
        if len(parts) < 2: continue
        lvl, tag, val = parts[0], parts[1], parts[2] if len(parts)>2 else ""

        if lvl == "0" and "INDI" in val:
            current_id = tag.replace("@", "")
            if not current_id.startswith('I'): current_id = 'I' + current_id
            individuals[current_id] = {"name": "findme", "famc": None, "fams": [], "sex": "U", "raw_name": ""}
            persons_data[current_id] = {'name': f'ID {current_id}', 'bdate': '', 'bplace': '', 'ddate': '', 'dplace': '', 'mdate': '', 'sources_count': 0, 'citations_count': 0, 'raw_note': ''}
            mode = None
        elif current_id and lvl != "0":
            if tag == "NAME" and lvl == "1":
                individuals[current_id]["raw_name"] = val
                n = val.replace("/", "").strip()
                individuals[current_id]["name"] = n if n.lower() not in ["unknown","living"] else "findme"
                persons_data[current_id]["name"] = n
            elif tag == "SEX" and lvl == "1":
                individuals[current_id]["sex"] = val.strip()
            elif tag == "FAMC" and lvl == "1":
                individuals[current_id]["famc"] = val.replace("@", "")
            elif tag == "FAMS" and lvl == "1":
                individuals[current_id]["fams"].append(val.replace("@", ""))
            elif tag == "BIRT": mode = "BIRT"
            elif tag == "DEAT": mode = "DEAT"
            elif tag == "NOTE" and lvl == "1":
                persons_data[current_id]['raw_note'] = persons_data[current_id].get('raw_note', '') + " " + val
                mode = "NOTE"
            elif tag == "CONT" and mode == "NOTE": persons_data[current_id]['raw_note'] += " " + val
            elif tag == "CONC" and mode == "NOTE": persons_data[current_id]['raw_note'] += val
            elif tag == "SOUR" and lvl == "1": persons_data[current_id]['sources_count'] += 1; mode = None
            elif tag == "SOUR" and lvl == "2": persons_data[current_id]['citations_count'] += 1
            elif tag == "DATE" and mode == "BIRT": persons_data[current_id]['bdate'] = val
            elif tag == "DATE" and mode == "DEAT": persons_data[current_id]['ddate'] = val
            elif tag == "PLAC" and mode == "BIRT": persons_data[current_id]['bplace'] = val
            elif tag == "PLAC" and mode == "DEAT": persons_data[current_id]['dplace'] = val
            elif lvl == "1": mode = None
        elif lvl == "0" and "FAM" in val:
            current_fam = tag.replace("@", "")
            families[current_fam] = {"husb": None, "wife": None, "mdate": ""}
            current_id = None
            mode = None
        elif current_fam and lvl != "0":
            if tag == "HUSB":
                husb_id = val.replace("@", "")
                if not husb_id.startswith('I'): husb_id = 'I' + husb_id
                families[current_fam]["husb"] = husb_id
            elif tag == "WIFE":
                wife_id = val.replace("@", "")
                if not wife_id.startswith('I'): wife_id = 'I' + wife_id
                families[current_fam]["wife"] = wife_id
            elif tag == "MARR": mode = "MARR"
            elif tag == "DATE" and mode == "MARR": families[current_fam]["mdate"] = val
            elif lvl == "1": mode = None

for fid, f in families.items():
    if f.get("mdate"):
        if f["husb"] and f["husb"] in persons_data: persons_data[f["husb"]]["mdate"] = f["mdate"]
        if f["wife"] and f["wife"] in persons_data: persons_data[f["wife"]]["mdate"] = f["mdate"]

# --- STEP 3.5: COMPUTATIONAL NAMING ENGINE (UNIVERSAL DEPLOYMENT) ---
print("    [+] Formatting Names (Applying Spanish/Née Math to ALL Nodes)...")

def get_sur(pid):
    if pid in override_dict and override_dict[pid]['surname']: return override_dict[pid]['surname']
    if not pid or pid not in individuals: return ""
    rn = individuals[pid].get("raw_name", "")
    if "/" in rn: return rn.split("/")[1].strip()
    return rn.split()[-1] if len(rn.split()) > 1 else ""

def get_giv(pid):
    if pid in override_dict and override_dict[pid]['given']: return override_dict[pid]['given']
    if not pid or pid not in individuals: return ""
    rn = individuals[pid].get("raw_name", "")
    if "/" in rn: giv = rn.split("/")[0].strip()
    else: giv = " ".join(rn.split()[:-1]) if len(rn.split()) > 1 else rn
    return re.sub(r'\d+', '', giv).strip()

for pid, p in individuals.items():
    if p["name"] == "findme": continue
    giv = get_giv(pid); sur = get_sur(pid)
    d_id = None; m_id = None
    famc = p.get("famc")
    if famc and famc in families:
        d_id = families[famc].get("husb"); m_id = families[famc].get("wife")

    d_sur = get_sur(d_id).split('-')[0].strip() if d_id else ""
    m_sur = get_sur(m_id).split('-')[0].strip() if m_id else ""
    maiden_sur = f"{d_sur}-{m_sur}" if d_sur and m_sur and d_sur!="findme" and m_sur!="findme" else d_sur if d_sur and d_sur!="findme" else m_sur if m_sur and m_sur!="findme" else ""

    sex = p.get("sex", "U"); h_sur = ""
    if sex == "F" and p.get("fams"):
        for fid in p["fams"]:
            if fid in families:
                h_id = families[fid].get("husb")
                h_s = get_sur(h_id).split('-')[0].strip() if h_id else ""
                if h_s and h_s != "findme": h_sur = h_s; break

    final_married_sur = h_sur if h_sur else sur
    if sex == "F" and maiden_sur and final_married_sur.lower() != maiden_sur.lower() and not giv.endswith(maiden_sur):
        display_name = f"{giv} {final_married_sur}, née {maiden_sur}".strip()
        p["gedcom_name"] = f"{giv} /{final_married_sur}/, née {maiden_sur}".strip()
    else:
        use_sur = f"{sur}-{m_sur}" if maiden_sur and m_sur and sur.lower() != d_sur.lower() and d_sur else maiden_sur if maiden_sur else sur
        display_name = f"{giv} {use_sur}".strip()
        p["gedcom_name"] = f"{giv} /{use_sur}/".strip()

    p["name"] = display_name
    if pid in persons_data: persons_data[pid]["name"] = display_name
    if pid in csv_auth: csv_auth[pid]["name"] = display_name

print("    ✅ Phase 2 Complete. GEDCOM structured and computational naming universally applied.")
```

```python
# @title ⚙️ [CELL 3C] The ANCHOR Math & Network Demographics
# 🛑 DOCTRINE LOCK: YATES DNA ANCHOR STUDY
# RULE: ID-TO-GEDCOM NAMING ONLY.
# RULE: CSV IS TRUTH (CSS/DOCS); GEDCOM IS MAP.
# RULE: GENERATOR-FIRST (NO SURGICAL SERVER PATCHES).
# RULE: NO PARTIAL BLOCKS. COMPILE FULL CELL.

print("="*60)
print("      [PHASE 3] EXECUTING ANCHOR MATH & DEMOGRAPHICS...")
print("="*60)

# --- STEP 4: LINEAGE EXTRACTION ---
print("    [+] Tracing Lineages on the Map...")
def get_parents(pid):
    if not pid or pid not in individuals: return None, None
    famc = individuals[pid]["famc"]
    if not famc or famc not in families: return None, None
    return families[famc]["husb"], families[famc]["wife"]

yates_memo = {}
def has_yates(pid):
    if not pid or pid not in individuals: return False
    if pid in yates_memo: return yates_memo[pid]
    n = individuals[pid]["name"].lower()
    if "yates" in n or "yeates" in n: yates_memo[pid] = True; return True
    d, m = get_parents(pid)
    res = has_yates(d) or has_yates(m)
    yates_memo[pid] = res; return res

def climb(start_id):
    curr = start_id; lin = []
    while curr:
        p = individuals.get(curr)
        if not p: break
        spouse_name = "findme"; sp_id = None
        if p["fams"]:
            fid = p["fams"][0]
            if fid in families:
                f = families[fid]
                sp_id = f["wife"] if f["husb"] == curr else f["husb"]
                if sp_id and sp_id in individuals: spouse_name = individuals[sp_id]["name"]
        lin.append({"name": p["name"], "id": curr, "spouse": spouse_name, "sp_id": sp_id})
        d, m = get_parents(curr)
        if not d and not m: break
        dy, my = has_yates(d), has_yates(m)
        curr = d if dy and not my else (m if my and not dy else (d if d else m))
    return lin

for td in master_vips:
    if td.get("id") and td["id"] in individuals:
        full = list(reversed(climb(td["id"])))
        td["lin_str"] = " -> ".join([x["name"] for x in full])
        td["pids"] = ",".join([x["id"].replace('I','') for x in full])
        td["name"] = individuals[td["id"]]["name"]

rows = []
for vol in volume_list:
    uid = vol["match_id"]
    td = csv_auth[vol["tester_id"]]

    if uid not in individuals: continue
    p = individuals[uid]
    lin = climb(uid)
    if not lin: continue
    full = list(reversed(lin))

    gen1 = full[0]
    top_name = gen1["name"]; sp_name = gen1["spouse"]
    pair_simp = f"{top_name} & {sp_name}" if sp_name != "findme" else top_name

    b_yr = re.search(r'\d{4}', persons_data[gen1['id']].get('bdate', ''))
    d_yr = re.search(r'\d{4}', persons_data[gen1['id']].get('ddate', ''))
    dates = f"({b_yr.group(0) if b_yr else 'findme'} - {d_yr.group(0) if d_yr else 'findme'})"

    dir_lbl = f"{top_name} {dates}" + (f" & {sp_name}" if sp_name != "findme" else "")

    rows.append({
        "Tester_Code": td['id'].replace('I',''),
        "Tester_Name": td['name'],
        "Tester_ID": "",
        "Tester_Display": td['name'],
        "Tester_Sort_Key": td['sort_key'],
        "Tester_Lineage": td['lin_str'],
        "Tester_Path_IDs": td['pids'],
        "Match_Name": p["name"],
        "Match_ID": uid,
        "cM": vol["cm"],
        "Match_Lineage": " -> ".join([pair_simp] + [x["name"] for x in full[1:]]),
        "Match_Path_IDs": ",".join([x["id"].replace('I','') for x in full]),
        "Authority_Directory_Label": dir_lbl
    })

df = pd.DataFrame(rows)
if df.empty:
    print("    ⚠️ Notice: 0 matches processed.")
    db_json = []; participant_scores = []; anc_data = {}; derived_branches = {}; node_reviews = {}
else:
    df.to_csv(CSV_DB, index=False, encoding="iso-8859-15", quoting=csv.QUOTE_ALL)
    print(f"    ✅ Matrix Complete: {len(df)} lines written.")

    df.rename(columns={
        "Authority_Directory_Label": "Dir_Label",
        "Tester_Code": "Kit_Code",
        "Match_Lineage": "Lineage",
        "Match_Path_IDs": "s_ids",
        "Tester_Display": "Kit_Name"
    }, inplace=True)

    df['search_ids'] = df['s_ids']
    df['search_names'] = df['Lineage'].astype(str).str.replace(' -> ', '|')
    df['t_names'] = df['Tester_Lineage'].astype(str).str.replace(' -> ', '|')
    df['t_ids'] = df['Tester_Path_IDs'].astype(str).str.replace(',', '|')

    # --- STEP 5: MATHEMATICAL ENGINE (ANCHOR/CSS) ---
    print("    [+] Calculating CSS & DOCS Math Matrices...")
    def clean_num(s): return re.sub(r'[^0-9]', '', str(s))
    def norm_log(val, cap): return min(1.0, math.log(1 + max(0, val or 0)) / math.log(1 + cap))

    id_to_kits = {}
    for _, r in df.iterrows():
        for i in [clean_num(x) for x in str(r['s_ids']).split(',') if x]:
            id_to_kits.setdefault(i, set()).add(r['Kit_Name'])

    participant_scores = []
    for p_name, grp in df.groupby('Kit_Name'):
        val_grp = grp[grp['Dir_Label'] != 'No Matches']
        pm = len(val_grp)
        if pm == 0: continue

        dc = val_grp['Dir_Label'].value_counts()
        hc_t, hc_2 = (int(dc.iloc[0]), int(dc.iloc[1])) if len(dc)>1 else (int(dc.iloc[0]), 0) if len(dc)>0 else (0,0)
        target_anc = str(dc.index[0]) if len(dc)>0 else "Unknown"

        hh, target_id = 0, None
        for _, r in val_grp.iterrows():
            for i in [clean_num(x) for x in str(r['s_ids']).split(',') if x]:
                if len(id_to_kits.get(i, set())) > hh: hh, target_id = len(id_to_kits[i]), i

        spine_ids = []; br = 0; tb = 0
        dom_branch_pct = 0; branch_bal = "0 : 0"; dist_support = "Standard Convergence"

        if target_id:
            tb = len(id_to_kits.get(target_id, set()))
            col = df[df['s_ids'].apply(lambda x: target_id in [clean_num(y) for y in str(x).split(',') if y])]
            branch_counts = {}
            for _, r in col.iterrows():
                ids = [clean_num(x) for x in str(r['s_ids']).split(',') if x]
                nms = str(r['search_names']).split('|')
                if target_id in ids:
                    idx = ids.index(target_id)
                    b_name = nms[idx+1].replace('findme', '?').replace('FINDME', '?').split(' (')[0].strip() if idx+1 < len(nms) else "Direct"
                    branch_counts[b_name] = branch_counts.get(b_name, 0) + 1

            br = len(branch_counts)
            spine_ids = [clean_num(x) for x in str(val_grp[val_grp['s_ids'].apply(lambda x: target_id in str(x))].iloc[0]['s_ids']).split(',') if x]

            if branch_counts:
                sorted_counts = sorted(branch_counts.values(), reverse=True)
                max_b = sorted_counts[0]
                tot_b = sum(sorted_counts)
                if tot_b > 0:
                    dom_branch_pct = round((max_b / tot_b) * 100)
                    rest_pct = 100 - dom_branch_pct
                    branch_bal = f"{dom_branch_pct} : {rest_pct}"
                    if dom_branch_pct >= 80: dist_support = "Concentrated Lineage"
                    elif dom_branch_pct <= 50 and br >= 3: dist_support = "Highly Distributed"
                    else: dist_support = "Standard Convergence"

        dr = float(hc_t / max(1, hc_2))
        w_sum = norm_log(pm, 150) + norm_log(hc_t, 100) + (norm_log(dr, 10)*1.5) + norm_log(tb, 40) + norm_log(hh, 150)
        w_sum += 2.0 * (1.0 if br>=6 else 0.85 if br==5 else 0.70 if br==4 else 0.50 if br==3 else 0.25 if br==2 else 0)
        st_str, st_val = ("PASS", 1.0) if pm>=15 and br>=3 and dr>=1.5 else ("PARTIAL", 0.85) if pm>=15 and br>=2 else ("FAIL", 0.6)
        css = float(100 * (w_sum / 7.5) * st_val)

        diag_log = []
        for pid in spine_ids:
            d = persons_data.get(pid, {})
            if d.get('sources_count',0) + d.get('citations_count',0) == 0: diag_log.append(f"Missing Sources: {d.get('name')} (I{pid})")

        docs_score = 88.5 if len(diag_log)==0 else max(35.0, 85.0 - (len(diag_log)*5))
        anchor = min(100.0, css)

        participant_scores.append({
            'pName': str(p_name), 'targetAnc': target_anc, 'targetID': str(target_id or ''),
            'PM': int(pm), 'HC_T': int(hc_t), 'DR': float(dr), 'BR': int(br), 'NS': int(hh),
            'ST_str': st_str, 'cssFinal': float(css), 'DOCS': float(docs_score), 'ANCHOR': float(anchor),
            'diagLog': diag_log, 'AX': 1, 'CC': 1.0, 'BC': 1.0, 'DC': 1.0, 'TP': 1.0,
            'isGroup': False, 'GD': len(spine_ids),
            'domBranchPct': dom_branch_pct, 'branchBal': branch_bal, 'distSupport': dist_support
        })

    anc_data = {}
    derived_branches = {}
    node_reviews = {}

    for lbl, grp in df.groupby('Dir_Label'):
        if str(lbl).strip() == 'No Matches' or not str(lbl).strip(): continue
        u_t = len(grp['Kit_Name'].unique())

        # --- PATH DEMOGRAPHICS LOGIC ---
        pure_pat = 0; pure_mat = 0; mixed = 0; unresolved = 0
        for _, r in grp.iterrows():
            ids = [clean_num(x) for x in str(r['s_ids']).split(',') if x]
            if len(ids) > 1:
                line_sexes = [individuals.get(f"I{iid}", {}).get('sex', 'U') for iid in ids[:-1]]
                if all(s == 'M' for s in line_sexes): pure_pat += 1
                elif all(s == 'F' for s in line_sexes): pure_mat += 1
                elif 'U' in line_sexes: unresolved += 1
                else: mixed += 1
            else:
                unresolved += 1

        anc_data[str(lbl)] = {
            "name": str(lbl),
            "matches": len(grp),
            "cm": int(grp['cM'].sum()),
            "badge": "Platinum" if len(grp)>=30 else "Gold",
            "integrity": min(100, (len(grp)*2) + (u_t*10)),
            "testers": u_t,
            "overall_case_confidence": "Moderate Confidence" if len(grp) >= 15 else "Low Confidence",
            "path_demographics": {
                "total_supporting": len(grp),
                "pure_matrilineal": pure_mat,
                "pure_patrilineal": pure_pat,
                "mixed_autosomal": mixed,
                "unresolved": unresolved
            }
        }

        branch_stats = {}
        gen_stats = {}
        total_matches = len(grp)

        for _, r in grp.iterrows():
            parts = str(r['Lineage']).split(' -> ')
            ids = [clean_num(x) for x in str(r['s_ids']).split(',') if x]

            if len(parts) > 1:
                b_name = parts[1].strip()
                b_id = ids[1] if len(ids) > 1 else None
            else:
                b_name = "Direct / Unknown"
                b_id = None

            if b_name not in branch_stats: branch_stats[b_name] = {'count': 0, 'id': b_id}
            branch_stats[b_name]['count'] += 1

            max_anc_idx = len(parts) - 1
            for g in range(2, max_anc_idx):
                p_name = parts[g-1].strip()
                node_name = parts[g].strip()
                node_id = ids[g] if len(ids) > g else None
                key = (p_name, node_name)

                if g not in gen_stats: gen_stats[g] = {}
                if key not in gen_stats[g]: gen_stats[g][key] = {'count': 0, 'id': node_id}
                gen_stats[g][key]['count'] += 1

        branch_list = []
        for b_name, stats in branch_stats.items():
            count = stats['count']
            pct = round((count / total_matches) * 100) if total_matches > 0 else 0

            if pct >= 60: contrib = "Dominant Contributor"
            elif pct >= 30: contrib = "Co-Dominant Contributor"
            elif pct >= 10: contrib = "Supporting Contributor"
            else: contrib = "Marginal Contributor"

            b_id = stats['id']
            has_vitals = False
            has_sources = False
            sex_val = "U"
            sex_marker = ""

            if b_id:
                clean_b_id = re.sub(r'[^0-9]', '', str(b_id))
                pd_obj = persons_data.get(clean_b_id) or persons_data.get("I" + clean_b_id) or {}
                if pd_obj.get('bdate') or pd_obj.get('ddate'): has_vitals = True
                if pd_obj.get('sources_count', 0) > 0 or pd_obj.get('citations_count', 0) > 0: has_sources = True

                # SURGICAL INJECTION: SEX DESIGNATOR
                ind_obj = individuals.get(f"I{clean_b_id}")
                if ind_obj:
                    s = ind_obj.get('sex', 'U')
                    if s in ['M', 'F']:
                        sex_val = s
                        sex_marker = f" [{s}]"

            if pct >= 40 and has_vitals and has_sources: supp = "Strong Support"
            elif pct >= 15 and has_vitals: supp = "Adequate Support"
            elif pct < 15 or not has_vitals: supp = "Thin Support"
            else: supp = "Unresolved Support"

            notes = f"{pct}% of network flow."
            if not has_vitals: notes += " Missing core vitals."
            if not has_sources: notes += " Lacks direct sources."

            if supp == "Thin Support": next_move = "Targeted documentary research."
            elif supp == "Strong Support": next_move = "Shift focus to Gen 2 nodes."
            else: next_move = "Expand collateral testing."

            branch_list.append({
                "derived_first_ancestor": f"{b_name}{sex_marker}",
                "relationship": "Child",
                "contribution_pattern": contrib,
                "support_condition": supp,
                "support_notes": notes,
                "next_move": next_move,
                "count": count,
                "node_sex": sex_val
            })

        branch_list.sort(key=lambda x: x['count'], reverse=True)
        derived_branches[str(lbl)] = branch_list

        active_b = len(branch_stats)
        anc_data[str(lbl)]["network_condition"] = f"Concentrated Lineage ({active_b} Active Branches)" if active_b <= 3 else f"Distributed Lineage ({active_b} Active Branches)"

        node_reviews_list = []
        for g, nodes in gen_stats.items():
            parent_totals = {}
            parent_max = {}

            for (p_name, b_name), stats in nodes.items():
                c = stats['count']
                parent_totals[p_name] = parent_totals.get(p_name, 0) + c
                if c > parent_max.get(p_name, 0): parent_max[p_name] = c

            for (p_name, b_name), stats in nodes.items():
                count = stats['count']
                pct = round((count / total_matches) * 100) if total_matches > 0 else 0

                if pct >= 60: contrib = "Dominant Contributor"
                elif pct >= 30: contrib = "Co-Dominant Contributor"
                elif pct >= 10: contrib = "Supporting Contributor"
                else: contrib = "Marginal Contributor"

                if count == parent_max.get(p_name, 0) and count > 0: rel = "Primary Descending Line"
                else: rel = "Collateral Sub-branch"

                b_id = stats['id']
                has_vitals = False
                has_sources = False
                sex_val = "U"

                if b_id:
                    clean_b_id = re.sub(r'[^0-9]', '', str(b_id))
                    pd_obj = persons_data.get(clean_b_id) or persons_data.get("I" + clean_b_id) or {}
                    if pd_obj.get('bdate') or pd_obj.get('ddate'): has_vitals = True
                    if pd_obj.get('sources_count', 0) > 0 or pd_obj.get('citations_count', 0) > 0: has_sources = True

                    ind_obj = individuals.get(f"I{clean_b_id}")
                    if ind_obj:
                        s = ind_obj.get('sex', 'U')
                        if s in ['M', 'F']: sex_val = s

                if pct >= 40 and has_vitals and has_sources: supp = "Strong Support"
                elif pct >= 15 and has_vitals: supp = "Adequate Support"
                elif pct < 15 or not has_vitals: supp = "Thin Support"
                else: supp = "Unresolved Support"

                notes = f"{pct}% of network flow."
                if not has_vitals: notes += " Missing core vitals."
                if not has_sources: notes += " Lacks direct sources."

                if supp == "Thin Support": next_move = "Targeted documentary research."
                elif supp == "Strong Support": next_move = f"Shift focus to Gen {g+1} nodes."
                else: next_move = "Expand collateral testing."

                node_reviews_list.append({
                    "generation": g,
                    "node_ancestor": b_name,
                    "lead_emerged_ancestor": str(lbl),
                    "parent_line": p_name,
                    "relationship": rel,
                    "contribution_pattern": contrib,
                    "support_condition": supp,
                    "support_notes": notes,
                    "next_move": next_move,
                    "count": count,
                    "node_sex": sex_val
                })

        node_reviews_list.sort(key=lambda x: (x['generation'], x['parent_line'], -x['count']))
        node_reviews[str(lbl)] = node_reviews_list

    # --- TRUNCATION FIX: BROKEN INTO ARRAYS ---
    export_cols = [
        'Dir_Label', 'Kit_Name', 'cM', 'Match_ID', 'Lineage',
        's_ids', 'search_names', 't_names', 't_ids', 'Tester_ID', 'Kit_Code'
    ]
    rename_dict = {
        'Dir_Label': 'ancestor', 'Kit_Name': 'participant', 'cM': 'cm',
        'Match_ID': 'id', 'Lineage': 'lineage', 'Tester_ID': 'tester_id',
        'Kit_Code': 'kit_code'
    }
    db_json = df[export_cols].rename(columns=rename_dict).to_dict(orient='records')

    print("    ✅ Phase 3 Complete. Network matrices and Demographics calculated (with Node Sex data layer).")
```

```python
# @title 🧬 [CELL 3F] Segment Extraction Utility
# 🛑 DOCTRINE LOCK: YATES DNA ANCHOR STUDY
# RULE: ID-TO-GEDCOM NAMING ONLY.
# RULE: CSV IS TRUTH.
# RULE: NO PARTIAL BLOCKS. COMPILE FULL CELL.

import os, re, glob
import pandas as pd

print("="*60)
print("      [CELL 3F] EXTRACTING SEGMENT DATA FROM GEDCOM NOTES...")
print("="*60)

KEY_MATCHES = "key_matches.csv"
OUTPUT_CSV = "match_segments.csv"
DATA_STORE_DIR = "yatesville.net/vault_yates/vault_data-store-inspect-inout"

if not os.path.exists(KEY_MATCHES):
    print(f"❌ ERROR: {KEY_MATCHES} not found on the factory floor.")
else:
    # --- STEP 1: LOAD TARGET MATCH IDs ---
    print(f"    [+] Loading Target Match Volume from {KEY_MATCHES}...")
    df_matches = pd.read_csv(KEY_MATCHES, encoding='utf-8-sig').fillna('')
    target_ids = set()

    for _, row in df_matches.iterrows():
        m_id = str(row.get('Match_ID', '')).strip().upper()
        if not m_id: continue
        if not m_id.startswith('I'): m_id = 'I' + m_id
        target_ids.add(m_id)

    print(f"        -> Locked {len(target_ids)} unique Match IDs for scanning.")

    # --- STEP 2: LOCALIZE GEDCOM ---
    ged_files = [f for f in glob.glob("*.ged") if "_processed" not in f.lower()]
    if not ged_files:
        print("❌ ERROR: No native GEDCOM found.")
    else:
        INPUT_GED = sorted(ged_files, key=lambda x: os.path.getmtime(x), reverse=True)[0]
        print(f"    [+] Scanning Native GEDCOM Map: {INPUT_GED}")

        # --- STEP 3: MINE NOTES & REGEX MATCH ---
        extracted_data = []
        current_id = None
        current_note = ""
        in_note = False

        # Helper function to process the note text
        def process_note(pid, text):
            if pid in target_ids:
                # Regex hunts for "X segment" or "X segments"
                # e.g., "73 cM across 3 segments" -> captures "3"
                match = re.search(r'(\d+)\s*segments?', text, re.IGNORECASE)
                if match:
                    extracted_data.append({"Match_ID": pid, "Segments": match.group(1)})
                else:
                    extracted_data.append({"Match_ID": pid, "Segments": "0"})

        with open(INPUT_GED, "r", encoding="utf-8", errors="replace") as f:
            for line in f:
                line = line.strip()
                parts = line.split(" ", 2)
                if len(parts) < 2: continue
                lvl, tag = parts[0], parts[1]
                val = parts[2] if len(parts) > 2 else ""

                if lvl == "0" and "INDI" in val:
                    # Process the previous person's accumulated note before moving on
                    if current_id and in_note:
                        process_note(current_id, current_note)

                    current_id = tag.replace("@", "")
                    if not current_id.startswith('I'): current_id = 'I' + current_id
                    current_note = ""
                    in_note = False

                elif current_id and lvl != "0":
                    if tag == "NOTE" and lvl == "1":
                        current_note += " " + val
                        in_note = True
                    elif tag == "CONT" and in_note:
                        current_note += " " + val
                    elif tag == "CONC" and in_note:
                        current_note += val
                    elif lvl == "1":
                        # If we hit another level 1 tag, the NOTE block is finished
                        if in_note:
                            process_note(current_id, current_note)
                            in_note = False

        # Catch the very last person in the file
        if current_id and in_note:
            process_note(current_id, current_note)

        # --- STEP 4: EXPORT & DEPLOY ---
        df_out = pd.DataFrame(extracted_data)

        if df_out.empty:
            print("    ⚠️ Notice: No segment data was extracted. Check note formatting.")
        else:
            # Ensure no duplicates if a match had multiple notes, keep the highest segment count
            df_out['Segments'] = pd.to_numeric(df_out['Segments'], errors='coerce').fillna(0).astype(int)
            df_out = df_out.sort_values('Segments', ascending=False).drop_duplicates(subset=['Match_ID'])

            df_out.to_csv(OUTPUT_CSV, index=False)
            print(f"    ✅ Extraction Complete: {len(df_out)} segment records captured.")

            # Vault Push (If SFTP is active)
            if 'sftp' in globals() and transport.is_active():
                try:
                    sftp.put(OUTPUT_CSV, f"{DATA_STORE_DIR}/{OUTPUT_CSV}")
                    print(f"    ✅ SUCCESS: {OUTPUT_CSV} successfully pushed to the Vault.")
                except Exception as e:
                    print(f"    ⚠️ SFTP Push Failed: {e}. (Saved locally)")
            else:
                print(f"    ✅ {OUTPUT_CSV} generated locally. (SFTP bypassed)")

print("\n[BIRD 2 SECURED] Segment extraction utility standing by.")
```

```python
# @title 🗺️ [UTILITY] SFTP Directory Deep Crawler (Target Corrected)
import stat
from google.colab import files

TARGET_DIR = "yatesville.net/vault_yates"

print("="*60)
print(f"      📡 Mapping the Layout of: {TARGET_DIR}")
print("="*60)

def sftp_tree(sftp_client, remote_path, file_handle, prefix=""):
    try:
        # listdir_attr pulls files and metadata in one shot
        items = sftp_client.listdir_attr(remote_path)
    except Exception as e:
        msg = f"{prefix} [⚠️ Access Denied / Error: {e}]"
        print(msg)
        file_handle.write(msg + "\n")
        return

    # Sort directories first, then files alphabetically
    items.sort(key=lambda x: (not stat.S_ISDIR(x.st_mode), x.filename.lower()))

    for i, item in enumerate(items):
        if item.filename in ('.', '..'):
            continue

        is_last = (i == len(items) - 1)
        connector = "└── " if is_last else "├── "
        child_prefix = prefix + ("    " if is_last else "│   ")

        if stat.S_ISDIR(item.st_mode):
            line = f"{prefix}{connector}📁 {item.filename}/"
            print(line)
            file_handle.write(line + "\n")
            # Recurse into directory
            sftp_tree(sftp_client, f"{remote_path}/{item.filename}", file_handle, child_prefix)
        else:
            # Format size
            size_bytes = item.st_size
            if size_bytes > 1024 * 1024: size_str = f"{size_bytes / (1024*1024):.2f} MB"
            elif size_bytes > 1024: size_str = f"{size_bytes / 1024:.2f} KB"
            else: size_str = f"{size_bytes} B"

            # Format time
            import time
            mod_time = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(item.st_mtime))

            line = f"{prefix}{connector}📄 {item.filename} ({size_str} | Mod: {mod_time})"
            print(line)
            file_handle.write(line + "\n")

# Execution
if 'sftp' not in globals():
    print("❌ ERROR: SFTP pipe is not open. Please run Cell 1 first.")
else:
    output_file = "vault_yates_map.txt"
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(f"📡 SFTP Scan: {TARGET_DIR}\n")
        f.write("="*60 + "\n")
        root_line = f"📁 {TARGET_DIR}/"
        print(root_line)
        f.write(root_line + "\n")

        # Trigger recursive crawl
        sftp_tree(sftp, TARGET_DIR, f)

    print("="*60)
    print(f"✅ Map saved to {output_file}. Triggering download...")
    try:
        files.download(output_file)
    except:
        print("⚠️ Could not auto-download. Please check the colab files pane.")
```

```python
# @title [CELL 7] The Time Machine (Vault Archive & Dropbox Sync)
def run_archiver():
    print("="*60)
    print("      [CELL 7] THE TIME MACHINE (Vault Archive & Sync)")
    print("="*60)

    import zipfile, os, pytz, json
    from datetime import datetime
    from google.colab import files
    try: from google.colab import userdata
    except ImportError: pass

    if 'sftp' not in globals():
        return print("❌ ERROR: SFTP pipe is not open. Please run Cell 1 first.")

    est = pytz.timezone('US/Eastern')
    timestamp = datetime.now(est).strftime("%Y-%m-%d_%H%M_%S")
    target_dir = 'site' if os.path.exists('site') else '.'

    # 1. GENERATE LEGACY README
    print("    [STEP 1] Generating Legacy Handover Protocol...")
    readme_content = f"""=========================================================
YATES ONS STUDY - LEGACY ENGINE & DATA ARCHIVE
Archive Date: {datetime.now(est).strftime("%B %d, %Y")}
=========================================================
TO THE FUTURE RESEARCHER:
You have inherited the complete, self-sustaining dataset and processing engine
for the Yates DNA Study. Everything required to rebuild and publish this
genealogical network is contained within this folder.
========================================================="""

    readme_name = "000_START_HERE_YATES_LEGACY.txt"
    with open(readme_name, "w", encoding="utf-8") as f: f.write(readme_content)

    # 2. GENERATE SITE SNAPSHOT JSON
    print("    [STEP 2] Generating Site Snapshot JSON...")
    snapshot_data = {}
    html_files = [f for f in os.listdir(target_dir) if f.lower().endswith(('.shtml', '.html'))]
    for f_name in html_files:
        try:
            with open(os.path.join(target_dir, f_name), 'r', encoding='utf-8') as fh:
                snapshot_data[f_name] = fh.read()
        except: pass

    snapshot_name = f"site_snapshot_{timestamp}.json"
    with open(snapshot_name, 'w', encoding='utf-8') as f: json.dump(snapshot_data, f)

    # 3. CREATE MASTER ZIP VAULT
    extensions = ('.csv', '.shtml', '.html', '.json', '.js', '.css', '.ged', '.ipynb')
    files_to_pack = [readme_name]

    for root_dir, dirs, files_in_dir in os.walk('.'):
        if '.config' in root_dir or 'sample_data' in root_dir: continue
        for f in files_in_dir:
            if f.lower().endswith(extensions) and f != snapshot_name:
                files_to_pack.append(os.path.join(root_dir, f))

    zip_name = f"Yates_Study_Legacy_Vault_{timestamp}.zip"
    print(f"    [STEP 3] Compressing {len(files_to_pack)} files into {zip_name}...")
    try:
        with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zf:
            for file in files_to_pack:
                arcname = file.replace('./', '') if file.startswith('./') else file
                zf.write(file, arcname)
        print(f"    ✅ Archive Created: {os.path.getsize(zip_name)/1024/1024:.2f} MB")
    except Exception as e:
        return print(f"    ❌ Compression Failed: {e}")

    # 4. SFTP UPLOAD TO VAULT ARCHIVE & ROLLING 10 CLEANUP
    print("\n    [STEP 4] Pushing to Vault Archive via SFTP...")
    ARCHIVE_DIR = "yatesville.net/vault_yates/vault_archive"
    try:
        sftp.put(zip_name, f"{ARCHIVE_DIR}/{zip_name}")
        print(f"    ✅ Uploaded to: {ARCHIVE_DIR}/{zip_name}")

        print("    [*] Running Vault Cleanup (Rolling 10 Policy)...")
        server_files = sftp.listdir(ARCHIVE_DIR)
        zip_backups = sorted([f for f in server_files if f.startswith("Yates_Study_Legacy_Vault_") and f.endswith(".zip")])

        zips_to_delete = zip_backups[:-10]
        if not zips_to_delete: print("      -> Archive is under 10 backups. Clean.")
        else:
            for old_zip in zips_to_delete:
                sftp.remove(f"{ARCHIVE_DIR}/{old_zip}")
                print(f"      🗑️ Incinerated old Vault ZIP: {old_zip}")
    except Exception as e: print(f"    ⚠️ Vault Archive Upload failed: {e}")

    # 5. DROPBOX SYNC
    print("\n    [STEP 5] Syncing to Dropbox Omni-Vault...")
    try:
        dbx_app_key = os.environ.get("DBX_APP_KEY") or userdata.get("DBX_APP_KEY")
        dbx_app_secret = os.environ.get("DBX_APP_SECRET") or userdata.get("DBX_APP_SECRET")
        dbx_refresh = os.environ.get("DBX_REFRESH_TOKEN") or userdata.get("DBX_REFRESH_TOKEN")

        if not dbx_refresh: print("    ❌ 'DBX_REFRESH_TOKEN' not found.")
        else:
            try: import dropbox
            except ImportError: os.system('pip install dropbox -q'); import dropbox

            dbx = dropbox.Dropbox(app_key=dbx_app_key, app_secret=dbx_app_secret, oauth2_refresh_token=dbx_refresh)
            folder_path = "/Yates_Study_Sync/archives"
            zip_target = f"{folder_path}/{zip_name}"

            with open(zip_name, "rb") as f: dbx.files_upload(f.read(), zip_target)
            print(f"    ✅ Dropbox Sync Success")

            # Dropbox Sweep
            result = dbx.files_list_folder(folder_path)
            dbx_files = [entry.name for entry in result.entries if isinstance(entry, dropbox.files.FileMetadata)]
            dbx_zips = sorted([f for f in dbx_files if f.startswith("Yates_Study_Legacy_Vault_") and f.endswith(".zip")])
            for old_zip in dbx_zips[:-10]:
                dbx.files_delete_v2(f"{folder_path}/{old_zip}")
    except Exception as e: print(f"    ❌ Dropbox Upload Failed: {e}")

    print("\n    [STEP 6] Triggering Local Download...")
    try: files.download(zip_name)
    except: pass

run_archiver()
```

```python
# @title 🖨️ [CELL 6] Report Generator Engine (Vault Architecture)
participant_search = "Ronald Yates-Carpenter" # @param {type:"string"}

import os, json, io, re
import pandas as pd
from datetime import datetime
from jinja2 import Environment, DictLoader

def make_cascade(lin_str):
    if not lin_str: return ""
    parts = [p.strip() for p in str(lin_str).split('->')]
    html = '<div style="text-align:left; font-size:13px; line-height:1.6; font-family: Georgia, serif; margin:4px 0;">'
    for i, p in enumerate(parts):
        pad = i * 15
        prfx = '' if i == 0 else '&uarr; '
        fw = 'font-weight:bold; color:#000;' if i == 0 else 'color:#444;'
        html += f'<div style="margin-left:{pad}px; {fw}">{prfx}{p}</div>'
    html += '</div>'
    return html

try:
    print("="*60)
    print(f"      🖨️ INITIATING LIVE REPORT ENGINE: '{participant_search}'")
    print("="*60)

    if 'sftp' not in globals():
        raise Exception("SFTP pipe is not open. Please run Cell 1 first.")

    # 1. LOAD THE CSV RELATIONAL BRIDGE
    roster_dict = {}
    if os.path.exists("key_participants.csv"):
        print("    [*] Loading key_participants.csv master roster...")
        df_participants = pd.read_csv("key_participants.csv")
        df_participants = df_participants.loc[:, ~df_participants.columns.str.contains('^Unnamed')]
        for index, row in df_participants.iterrows():
            gedcom_id = str(row['Id_Key']).strip().upper()
            if gedcom_id and gedcom_id != 'NAN':
                roster_dict[gedcom_id] = {
                    "display_name": str(row['Display_Name']).strip(),
                    "study_id": str(row['Numeric_assigned']).strip()
                }

    # 2. Fetch the Database LOCALLY
    if not os.path.exists("compiled_database.json"):
        raise FileNotFoundError("compiled_database.json is missing! Please run Cell 3 first.")

    with open("compiled_database.json", "r", encoding="utf-8") as f:
        master_data = json.load(f)

    precomputed = master_data.get("PRECOMPUTED", [])
    raw_db = master_data.get("DB", [])

    # 3. Find the Target Participant
    search_lower = participant_search.strip().lower()
    target_subject = None

    # [FIX APPLIED] Pure Passthrough: Removed .split('[')[0] to preserve original strings
    for p in precomputed:
        p_name_clean = str(p.get('pName', '')).strip().lower()
        p_code = str(p.get('pCode', '')).lower()
        if search_lower in p_name_clean or search_lower in p_code:
            target_subject = p
            break

    if not target_subject:
        raise ValueError(f"Participant '{participant_search}' not found in database.")

    # [FIX APPLIED] Pure Passthrough here as well
    raw_p_name = str(target_subject.get('pName', 'Unknown')).strip()
    p_code = str(target_subject.get('pCode', '')).upper()

    # ✨ CSV BRIDGE: Override raw name with clean Display Name
    display_p_name = roster_dict.get(p_code, {}).get("display_name", raw_p_name)
    display_study_id = roster_dict.get(p_code, {}).get("study_id", p_code)

    print(f"    [+] Locked Subject: {display_p_name} [{display_study_id}]")

    print("    [*] Fetching reporting templates from Vault Reporting...")

    template_dict = {}
    plates_dict = {}

    def fetch_folder_to_dict(sftp_path, prefix, target_dict, is_plate=False):
        try:
            for item in sftp.listdir_attr(sftp_path):
                f = item.filename
                if f.endswith('.html'):
                    remote_file = f"{sftp_path}/{f}"
                    local_tmp = f"temp_{f}"
                    sftp.get(remote_file, local_tmp)
                    with open(local_tmp, 'r', encoding='utf-8') as tmp_f:
                        content = tmp_f.read()
                    os.remove(local_tmp)

                    if is_plate: target_dict[f.replace('.html', '')] = content
                    else: target_dict[f"{prefix}{f}"] = content
        except Exception as e:
            print(f"    [⚠️] Failed fetching from {sftp_path}: {e}")

    REPORT_BASE = "yatesville.net/vault_yates/vault_reporting"
    fetch_folder_to_dict(f"{REPORT_BASE}/templates", "reporting/templates/", template_dict)
    fetch_folder_to_dict(f"{REPORT_BASE}/partials", "reporting/partials/", template_dict)
    fetch_folder_to_dict(f"{REPORT_BASE}/plates", "", plates_dict, is_plate=True)

    env = Environment(loader=DictLoader(template_dict))
    master_tmpl = env.get_template("reporting/templates/tmpl_report_master.html")

    # 4. Extract Top Matches
    # [FIX APPLIED] Retain hyphens and Spanish accents during strict match lookup
    p_name_clean_for_match = re.sub(r'[^a-z0-9 \u00C0-\u024F\-]', '', raw_p_name.lower())
    my_matches = []
    name_parts = p_name_clean_for_match.split()

    for m in raw_db:
        m_name = str(m.get('Participant_Name', m.get('Tester_Display', ''))).lower()
        m_code = str(m.get('Participant_Code', m.get('Tester_Code', ''))).lower()

        code_match = (p_code and p_code.lower() in m_code and len(p_code) > 2)
        name_match = (len(name_parts) > 0 and all(part in m_name for part in name_parts))

        if code_match or name_match:
            raw_label = m.get('Dir_Label', m.get('Authority_Directory_Label', ''))
            match_id = str(m.get('search_ids', m.get('id', ''))).upper()

            if raw_label and raw_label != 'No Matches':
                clean_label = roster_dict.get(match_id, {}).get("display_name", raw_label)
                cm_val = float(re.sub(r'[^\d.]', '', str(m.get('cM', '0')))) if str(m.get('cM', '0')) else 0.0
                my_matches.append({
                    "name": clean_label,
                    "cm": cm_val,
                    "path": make_cascade(m.get('Lineage', m.get('Match_Lineage', '')))
                })

    my_matches.sort(key=lambda x: x['cm'], reverse=True)
    top_matches = my_matches[:10]

    # 5. Define Holistic Matrix
    status_str = target_subject.get("ST_str", "EVAL")
    badge_class = "badge-high" if status_str == "PASS" else "badge-low" if status_str == "FAIL" else "badge-medium"
    conf_level = "High" if status_str == "PASS" else "Low" if status_str == "FAIL" else "Evaluating"

    holistic_headers = ["Layer", "Status", "Confidence", "Notes"]
    holistic_row = [
        "Autosomal (CSS)",
        f"<span class='badge {badge_class}'>V2 {status_str}</span>",
        conf_level,
        f"Verified via {target_subject.get('PM', 0)} proper matches"
    ]

    report_payload = {
        "report_title": "Forensic DNA Validation Report",
        "subject_type": "individual",
        "subject_name": display_p_name,
        "subject_id": display_study_id,
        "report_date": datetime.now().strftime("%B %d, %Y"),
        "study_name": "Yates DNA ANCHOR Study",
        "study_confidence_status": "Confirmed" if status_str == "PASS" else "Speculative" if status_str == "FAIL" else "Evaluating",
        "css_score": round(float(target_subject.get("cssFinal", 0)), 2) if target_subject.get("cssFinal") else "-",
        "emerged_node": target_subject.get("targetAnc", "Unknown"),
        "pm": target_subject.get("PM", "-"),
        "hc_t": target_subject.get("HC_T", "-"),
        "dr": target_subject.get("DR", "-"),
        "br": target_subject.get("BR", "-"),
        "ns": round(float(target_subject.get("NS", 0)), 1) if target_subject.get("NS") else "-",
        "holistic_matrix_headers": holistic_headers,
        "holistic_matrix_row": holistic_row,
        "convergence_summary": f"<p style='margin:0; font-size:15px; color:#444;'>Based on the aggregation of <strong>{target_subject.get('PM', 0)}</strong> triangulated cross-branch matches generating a total volume of <strong>{round(float(target_subject.get('NS', 0)), 1) if target_subject.get('NS') else 0} cM</strong>, this participant's individual DNA evidence heavily converges on the identified historical node.</p>",
        "supporting_kits": top_matches,
        "appendix_enabled": False
    }

    report_payload.update(plates_dict)

    print("    [+] Assembling templates and injecting data...")
    final_html = master_tmpl.render(**report_payload)

    # Make the filename strictly safe for URL/OS
    safe_name = re.sub(r'[^a-zA-Z0-9]', '_', display_p_name)
    output_filename = f"report_individual_{safe_name}.html"

    STAGING_DIR = f"{REPORT_BASE}/staging/generated_html"
    print(f"    [+] Publishing to {STAGING_DIR}/{output_filename}...")

    with open(output_filename, "w", encoding="utf-8") as f:
        f.write(final_html)

    sftp.put(output_filename, f"{STAGING_DIR}/{output_filename}")
    os.remove(output_filename)

    print("="*60)
    print(f"✅ REPORT GENERATED AND VAULTED: {display_p_name}")
    print(f"🔗 https://yatesville.net/vault_yates/vault_reporting/staging/generated_html/{output_filename}")

except Exception as e:
    print(f"\n❌ REPORT ENGINE ERROR: {e}")
```

```python
# @title [ISOLATED EXTRACTION] The Deep-Scan Crawler
import os
import stat
from IPython.display import HTML, display

if 'sftp' not in globals():
    print("❌ STOPPING: SFTP connection not found. Run your connection block first.")
else:
    print("📡 Deploying Deep-Scan Crawler to locate the files natively. This may take a few seconds...")

    target_filenames = ["plate_test_lab.html", "tmpl_theory_page.html"]
    found_paths = []

    def crawl_sftp(current_path, depth=0):
        # Limit depth to prevent runaway loops in massive directories
        if depth > 7: return
        try:
            for attr in sftp.listdir_attr(current_path):
                filename = attr.filename
                if filename in [".", ".."]: continue

                full_path = current_path + "/" + filename

                if stat.S_ISDIR(attr.st_mode):
                    crawl_sftp(full_path, depth + 1)
                elif filename in target_filenames:
                    found_paths.append(full_path)
                    print(f"   🔍 FOUND NATIVE PATH: {full_path}")
        except Exception as e:
            pass

    # Start from the known working root
    crawl_sftp("yatesville.net/vault_yates")

    print("-" * 60)

    if not found_paths:
        print("❌ CRAWL COMPLETE: The files completely evaded the scanner. They may have been fully deleted from the server.")
    else:
        print(f"✅ CRAWL COMPLETE. Found {len(found_paths)} targets. Extracting now...")

        for path in found_paths:
            local_tmp = "temp_crawl_extract.html"
            try:
                sftp.get(path, local_tmp)
                with open(local_tmp, 'r', encoding='utf-8', errors='replace') as f:
                    raw_html = f.read()
                os.remove(local_tmp)

                print(f"➔ RENDERED: {path}")
                display(HTML(f"<div style='border:2px solid #00838f; padding:15px; margin: 10px 0; background: #e0f7fa; color: black; max-height: 500px; overflow-y: auto;'><strong>{path}</strong><hr><br>{raw_html}</div>"))
            except Exception as e:
                print(f"❌ Failed to extract {path}: {e}")
```
